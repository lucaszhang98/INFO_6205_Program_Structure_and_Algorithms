Q1. Arrange the following functions in increasing order of growth:
1.	log(n2)
2.	2(2n)
3.	3n3 + 2n2
4.	n4

6.	n2 + n
7.	10n
8.	log(n3)
9.	n1/3
10.	 n! * 2n
Solution:
1.	log(n2) is a logarithmic function. The exponent 2 in log(n2) doesn't affect the order of growth, so it's Θ(log(n)).
2.	2(2n) is an exponential function and it's Θ(2n).
3.	3n3 + 2n2 is a polynomial function. It has the exponent (n3), so it's Θ(n3).
4.	n4 is also a polynomial function, and it's Θ(n4).
5.	sqrt(n) + log(n) combines a square root term and a logarithmic term. Both terms grow slower than the polynomial terms, so this function grows slower than any polynomial but faster than logarithmic or constant terms. So, it's Θ(sqrt(n)).
6.	n2 + n is a polynomial function with an exponent of n2, which falls between n3 and n4 in terms of growth. So, it's Θ(n2).
7.	10n is another exponential function, and it grows extremely fast. It's Θ(10n).
8.	log(n3) is a logarithmic function. The exponent 3 in log(n^3) doesn't affect the order of growth, so it's still Θ(log(n)).
9.	n1/3 is a polynomial function with an exponent of 1/3. It grows slower than n2 but faster than linear functions, so it's Θ(n1/3).
10.	 n! * 2n involves a factorial and an exponential term. Factorials grow extremely fast, but exponentials grow even faster. So, this function grows faster than any of the previous functions and is Θ (n! * 2n).
Now, arrange these functions in increasing order of growth:
5 < 1 < 8 < 9 < 6 < 3 < 4 < 2 < 7 < 10

Q3. Following a catastrophic natural disaster, multiple disaster-stricken locations require assistance from various relief organizations. Unfortunately, there is often a disconnect between the available resources and the needs of these locations. Your task is to develop an algorithm to efficiently allocate relief resources to different disaster-stricken areas, considering the preferences of both the relief organizations and the affected locations.
Here's what you have:
•	A list of disaster-stricken areas, each with a unique number.
•	For each area, you know how much of each resource it needs, like food, water, medical supplies, and shelter.
•	A group of volunteers, each with their own preferences for the areas based on expertise and resource availability. Higher values mean stronger preferences.
•	Constraints that specify the minimum and maximum amount of each resource type that can be allocated to each area.
Your goal is to create an allocation plan that satisfies the following conditions:
•	Each disaster-stricken location must receive a specific amount of each resource type within the specified constraints.
•	The allocation plan must be stable, meaning that no relief organization and location would prefer each other over their current allocations.
•	Relief organizations must be allocated to locations in a way that maximizes the overall satisfaction. The satisfaction of a relief organization with a location is the sum of their preferences for that location's expertise and available resources.
Solution:
1.	While there are unassigned volunteers:
•	Pick the next unassigned volunteer (v1).
•	For v1, find the area (a1) that maximizes their preference score, considering expertise and available resources.
•	If a1 is unassigned:
•	Assign v1 to a1.
•	Update the available resources in a1 based on the assigned volunteer.
•	If a1 is already assigned to another volunteer (v2):
•	Compare v1's preference for a1 and v2's preference for a1.
•	If v1 prefers a1 over v2's assignment:
•	Reassign a1 from v2 to v1.
•	Update the available resources in a1 accordingly.
•	Mark v2 as unassigned.
•	Otherwise, v1 remains unassigned.
2.	Repeat step 2 until all volunteers are assigned or no further improvements can be made.

The statement is True. In every stable matching for this scenario, the pair (Alice, XYZ) will always belong to the matching.

Q9. The NBA is changing its playoff format using the Gale-Shapley matching algorithm. Rather than Western Conference teams playing Western Conference teams; the eight Western Conference teams will be matched against the eight Eastern Conference teams using the Gale-Shapley matching algorithm. Further social media will be used to ask fans, media, players and coaches to create a ranking of which teams they most like to face.
C. Solution:
	The Gale-Shapley algorithm is used to create a single round of stable matching. Applying it repeatedly in each round does not guarantee overall stability over multiple rounds.
	Team and individual (fans, media, players and coaches) preferences can change from round to round. This means that a match that may be stable in one round may be unstable in the next round.
D. Solution:
The Gale-Shapley matching algorithm is designed to create stable matches in a scenario where two sets of agents with preferences are trying to be matched. In the context of the NBA playoff format change, where teams are ranked by fans, media, players, and coaches, the Gale-Shapley algorithm could still be used to create stable matches.
while the Gale-Shapley algorithm can still create stable matches in this context, the definition of "stability" may be more subjective and related to fan and team preferences rather than the mathematical stability of the algorithm itself.


```python
#A
def gale_shapley(western_prefs, eastern_prefs):
    eastern_matches = {}
    western_matches = {}
    free_eastern_teams = list(eastern_prefs.keys())
    
    while free_eastern_teams:
        eastern_team = free_eastern_teams.pop(0)
        eastern_preferences = eastern_prefs[eastern_team]
        
        for western_team in eastern_preferences:
            current_match = eastern_matches.get(eastern_team)
            
            if current_match is None:
                eastern_matches[eastern_team] = western_team
                western_matches[western_team] = eastern_team
                break
            else:
                current_rank = western_prefs[western_team].index(eastern_team)
                new_rank = western_prefs[western_team].index(current_match)
                
                if current_rank < new_rank:
                    break
                else:
                    eastern_matches[eastern_team] = western_team
                    western_matches[western_team] = eastern_team
                    free_eastern_teams.append(current_match)
    
    return western_matches

western_prefs = {
    "Nuggets": ["Celtics", "76ers", "Cavaliers", "Knicks", "Nets", "Hawks", "Heat", "Bucks"],
    "Grizzlies": ["76ers", "Cavaliers", "Knicks", "Nets", "Hawks", "Heat", "Bucks", "Celtics"],
    "Kings": ["Cavaliers", "Knicks", "Nets", "Hawks", "Heat", "Bucks", "Celtics", "76ers"],
    "Suns": ["Knicks", "Nets", "Hawks", "Heat", "Bucks", "Celtics", "76ers", "Cavaliers"],
    "Clippers": ["Nets", "Hawks", "Heat", "Bucks", "Celtics", "76ers", "Cavaliers", "Knicks"],
    "Warriors": ["Hawks", "Heat", "Bucks", "Celtics", "76ers", "Cavaliers", "Knicks", "Nets"],
    "Lakers": ["Heat", "Bucks", "Celtics", "76ers", "Cavaliers", "Knicks", "Nets", "Hawks"],
    "Timberwolves": ["Bucks", "Celtics", "76ers", "Cavaliers", "Knicks", "Nets", "Hawks", "Heat"]
}

eastern_prefs = {
    "Bucks": ["Grizzlies", "Kings", "Suns", "Clippers", "Warriors", "Lakers", "Timberwolves", "Nuggets"],
    "Celtics": ["Kings", "Suns", "Clippers", "Warriors", "Lakers", "Timberwolves", "Nuggets", "Grizzlies"],
    "76ers": ["Suns", "Clippers", "Warriors", "Lakers", "Timberwolves", "Nuggets", "Grizzlies", "Kings"],
    "Cavaliers": ["Clippers", "Warriors", "Lakers", "Timberwolves", "Nuggets", "Grizzlies", "Kings", "Suns"],
    "Knicks": ["Warriors", "Lakers", "Timberwolves", "Nuggets", "Grizzlies", "Kings", "Suns", "Clippers"],
    "Nets": ["Lakers", "Timberwolves", "Nuggets", "Grizzlies", "Kings", "Suns", "Clippers", "Warriors"],
    "Hawks": ["Timberwolves", "Nuggets", "Grizzlies", "Kings", "Suns", "Clippers", "Warriors", "Lakers"],
    "Heat": ["Nuggets", "Grizzlies", "Kings", "Suns", "Clippers", "Warriors", "Lakers", "Timberwolves"]
}

matches = gale_shapley(western_prefs, eastern_prefs)

for western_team, eastern_team in matches.items():
    print(f"Western Team {western_team} matched with Eastern Team {eastern_team}")
```

```python
#B
import random

num_teams = 8

eastern_teams = [list(range(num_teams)) for _ in range(num_teams)]
western_teams = [list(range(num_teams)) for _ in range(num_teams)]

num_iterations = 1000
stable_matches = 0

for _ in range(num_iterations):
    for i in range(num_teams):
        random.shuffle(eastern_teams[i])
        random.shuffle(western_teams[i])
    
    matches = [-1] * num_teams  # Initialize matches as -1 for all teams
    while -1 in matches:
        for i in range(num_teams):
            if matches[i] == -1:
                for j in eastern_teams[i]:
                    if matches[j] == -1:
                        matches[i] = j
                        matches[j] = i
                        break
    
    # Check if the matches are stable
    is_stable = True
    for i in range(num_teams):
        for j in range(i + 1, num_teams):
            if eastern_teams[i].index(matches[i]) > eastern_teams[i].index(matches[j]) and western_teams[j].index(i) > western_teams[j].index(matches[j]):
                is_stable = False
                break
    
    if is_stable:
        stable_matches += 1

percentage_stable = (stable_matches / num_iterations) * 100

print(f"Percentage of stable playoff matches: {percentage_stable}%")
```

```python
#E
import random
import time

def generate_preference_lists(num_teams):
    eastern_teams = [list(range(num_teams)) for _ in range(num_teams)]
    western_teams = [list(range(num_teams)) for _ in range(num_teams)]
    return eastern_teams, western_teams

def gale_shapley(eastern_teams, western_teams):
    num_teams = len(eastern_teams)
    matches = [-1] * num_teams
    while -1 in matches:
        for i in range(num_teams):
            if matches[i] == -1:
                for j in eastern_teams[i]:
                    if matches[j] == -1:
                        matches[i] = j
                        matches[j] = i
                        break
    return matches

def is_stable_match(eastern_teams, western_teams, matches):
    num_teams = len(eastern_teams)
    for i in range(num_teams):
        for j in range(i + 1, num_teams):
            if eastern_teams[i].index(matches[i]) > eastern_teams[i].index(matches[j]) and western_teams[j].index(i) > western_teams[j].index(matches[j]):
                return False
    return True

list_sizes = [8, 16, 32, 64, 128, 256]

for num_teams in list_sizes:
    eastern_teams, western_teams = generate_preference_lists(num_teams)
    
    start_time = time.perf_counter()
    matches = gale_shapley(eastern_teams, western_teams)
    end_time = time.perf_counter()
    
    if not is_stable_match(eastern_teams, western_teams, matches):
        print(f"Unstable match found for {num_teams} teams.")
    
    execution_time = end_time - start_time
    print(f"Execution time for {num_teams} teams: {execution_time:.8f} seconds")
```
