<h4>Q1. Comparing Ford-Fulkerson and Preflow-Push Algorithms for Maximum Flow (15 points)</h4>

<h5> Problem Statement:</h5>
You are given a weighted directed graph with a source node 'S' and a sink node 'T.' The goal is to find the maximum flow from 'S' to 'T' while respecting the capacity constraints on each edge. There are two algorithms commonly used for solving this problem: the Ford-Fulkerson algorithm and the Preflow-Push algorithm.
<br><br>
a. Explain how each of these algorithms works to find the maximum flow in a flow network. Compare and contrast their approaches and discuss scenarios where one might be preferred over the others. (10 points)
<br><br>
b. Additionally, provide a brief example to illustrate the application of one of these algorithms in a specific network flow problem. (5 points)
<br><br>
<strong>Solution:</strong>

Part A:
<h5><strong>Ford-Fulkerson Algorithm:</strong></h5>
<strong>Operation:</strong> The Ford-Fulkerson algorithm starts with an initial flow of zero and iteratively finds augmenting paths from the source 'S' to the sink 'T' in the residual graph. It augments the flow along these paths until no more augmenting paths can be found.
<br>
<strong>Residual Graph:</strong> The residual graph initially mirrors the original graph but adjusts as flow is pushed through. It contains the residual capacities, which may change as the flow increases.
<br>
<strong>Termination:</strong> The Ford-Fulkerson algorithm does not guarantee termination for graphs with real-valued capacities. In such cases, it may result in an infinite loop.
<br>
<strong>Integer and Real-Valued Capacities:</strong> It can handle graphs with both integer and real-valued capacities.
<br>
<strong>Preference:</strong> It is a flexible approach, but it may not be the most efficient option for dense graphs.
<br>
<h5><strong>Preflow-Push Algorithm:</h5></strong>
<strong>Operation:</strong> The Preflow-Push algorithm maintains a "preflow" from 'S' to 'T' and then iteratively pushes flow from nodes with excess preflow to nodes with deficits, while also performing "relabel" operations to optimize the flow.
<br>
<strong>Efficiency:</strong> It includes the "relabel" operation to improve efficiency by adjusting the preflow and reducing the number of flow augmentations.
<br>
<strong>Integer and Real-Valued Capacities:</strong> The Preflow-Push algorithm handles both integer and real-valued capacities efficiently.
<br>
<strong>Preference:</strong> It is often preferred for dense graphs due to its faster convergence and efficient handling of various capacities.
<br>
<h5><strong>Comparison:</strong></h5>
- Ford-Fulkerson is a general approach to finding maximum flow, while Preflow-Push combines the benefits of both Ford-Fulkerson and Edmonds-Karp, making it more efficient for many scenarios.
<br>
- Preflow-Push is often preferred for dense graphs, thanks to its faster convergence and efficient handling of various capacities.
<br>
- Ford-Fulkerson is a flexible approach but may not guarantee termination for graphs with real-valued capacities. It's more suitable for exploring different flow augmentation strategies.

Part B:

<strong>Example: Ford-Fulkerson Algorithm in Network Flow </strong>
<br><br>
Consider a transportation network with the following characteristics:
<br><br>
Source 'S' with a supply of 10 units.
<br><br>
Sink 'T' with a demand of 10 units.

<br><br>

Edge from Node C to 'T': Capacity = 7 units<br>
<br>
The goal is to find the maximum flow from 'S' to 'T' while respecting capacity constraints. We'll apply the Ford-Fulkerson algorithm to solve this problem:
<br><br>
Initialize the flow on all edges to 0.<br>
Find an augmenting path from 'S' to 'T' in the residual graph (initially the same as the original graph).<br>
Update the flow along the augmenting path, increasing the flow value by the smallest capacity on the path.<br>
Update the residual graph by subtracting the flow from forward edges and adding the flow to reverse edges.<br>
Repeat steps 2-4 until no more augmenting paths can be found.<br>
<br>
In this example, the Ford-Fulkerson algorithm might start by finding an augmenting path 'S -> A -> B -> T' with a minimum capacity of 4 units. The flow is updated, and the residual graph is adjusted. Then, another augmenting path 'S -> A -> C -> T' with a minimum capacity of 5 units is found and updated. The algorithm continues until no more augmenting paths can be found.
<br><br>
The maximum flow achieved while respecting capacity constraints in this network will be 10 units, matching the demand at 'T'. This example demonstrates how the Ford-Fulkerson algorithm can efficiently solve network flow problems by finding augmenting paths and updating flows in the graph.

<h4>Q2. Flight Delay Investigation (10 Points)</h4>

<h5> Problem Statement:</h5>
You are investigating flight delays at a major airport. You have access to data on the flights, their departure times, and potential delays. Your goal is to identify the critical flights that, if delayed, would cause a cascading effect leading to maximum overall delay in the airport's flight schedule.
<br><br>
<strong>Input Format:</strong>
<br><br>
flight_data (List of Tuples): A list of flight information, where each tuple is in the format (flight_id, departure_time, potential_delay). The flight_id is a unique identifier for the flight, departure_time is the scheduled departure time, and potential_delay is the maximum potential delay in minutes.
<br><br>
flight_connections (List of Tuples): A list of flight connections, where each tuple is in the format (source_flight_id, target_flight_id). It represents the connection between flights where the delay of the source flight can affect the target flight.
<br><br>
<strong>Output Format:</strong>
<br><br>
critical_flights (List of Integers): A list of flight_ids representing the critical flights that, if delayed, would lead to the maximum overall delay in the airport's flight schedule.
<br><br>
<strong>Sample Input:</strong>
<br>
<br>
flight_data = [(1, 300, 15), (2, 310, 10), (3, 320, 5), (4, 330, 8), (5, 340, 12)]<br><br>
flight_connections = [(1, 2), (2, 3), (3, 4), (4, 5)]
<br><br>
<strong>Sample Output:</strong>
<br><br>
Critical Flights: [1, 2, 3, 4, 5]
<br><br>
<strong>Constraints:</strong>
<br><br>
1. The departure_time is in minutes past midnight.<br>
2. The number of flights is at most 100.<br>
3. The potential_delay for each flight is a positive integer.<br>
4. The number of flight connections is at most 100.
<br><br>
<strong>Note:</strong>
<br><br>
In this problem, you need to use the Bellman-Ford algorithm to identify the critical flights that, if delayed, would lead to the maximum overall delay in the airport's flight schedule.
<br><br>
<strong>Solution/PseudoCode:</strong>

```python
BellmanFordCriticalFlights(flight_data, flight_connections)
    Initialize distances as an array with size equal to the number of flights, initialized to infinity
    Initialize parent_flights as an array with size equal to the number of flights, initialized to None
    Initialize critical_flights as an empty list

    Set the departure_time of the first flight as time=0
    distances[0] = 0  # The first flight has no potential delay

    # Relax edges using the Bellman-Ford algorithm
    for i from 1 to number of flights - 1:
        for each flight in flight_data:
            for each connection in flight_connections:
                source_flight_id, target_flight_id = connection
                source_flight = flight_data[source_flight_id]
                target_flight = flight_data[target_flight_id

                potential_delay = source_flight.potential_delay
                if distances[source_flight_id] + potential_delay < distances[target_flight_id]:
                    distances[target_flight_id] = distances[source_flight_id] + potential_delay
                    parent_flights[target_flight_id] = source_flight_id

    # Detect negative cycles
    for each connection in flight_connections:
        source_flight_id, target_flight_id = connection
        source_flight = flight_data[source_flight_id]
        target_flight = flight_data[target_flight_id

        potential_delay = source_flight.potential_delay
        if distances[source_flight_id] + potential_delay < distances[target_flight_id]:
            # Negative cycle detected, mark critical flights
            current_flight_id = target_flight_id
            while current_flight_id is not None:
                critical_flights.append(current_flight_id)
                current_flight_id = parent_flights[current_flight_id]

    return critical_flights
```

<strong>Justification:</strong>
<br><br>
The solution employs the Bellman-Ford algorithm to analyze potential flight delays and determine critical flights that could lead to maximum overall delay in the airport's flight schedule. It starts by initializing arrays to track distances and parent flights, then iteratively relaxes edges considering potential delays. The algorithm detects negative cycles, signifying situations where certain flights, if delayed, would trigger a cascading effect causing significant overall delays. The critical flights are identified by backtracking through the parent flights within negative cycles. These critical flights are the ones most pivotal in affecting the airport's overall delay scenario, making them essential to address for efficient flight management.

<strong>Solution:</strong>

To efficiently schedule the tasks and maximize the total importance, I would employ a dynamic programming approach for the Weighted Interval Scheduling problem. Here are the steps:
<br><br>
Sort the tasks: I would start by sorting the tasks based on their end times in ascending order. This sorting step ensures that we work with a list of tasks in non-decreasing order of their end times.
<br><br>
Initialize a dynamic programming table: I would set up a table to keep track of the maximum total importance achievable up to each task's end time. The table will have entries corresponding to each task.
<br><br>
Dynamic Programming: I'd iterate through the sorted tasks. For each task, I'd consider two options:
<br><br>
- Include the task: If I include the current task, I'd find the task with the latest end time that doesn't overlap with it. This means finding the task with the largest end time that has a start time less than or equal to the current task's end time. I'd add the weight of the current task to the maximum total importance achievable up to that non-overlapping task.
<br><br>
- Exclude the task: Alternatively, I'd consider the maximum total importance achieved without including the current task.

Fill the dynamic programming table: By iteratively considering the options of including or excluding each task, I'd populate the dynamic programming table. The entry at the last task's end time will represent the maximum total importance achievable for the entire project without any overlapping tasks.
<br><br>
Backtracking: If necessary, I could backtrack through the dynamic programming table to determine which tasks were selected to achieve the maximum total importance.
<br><br>
This dynamic programming approach ensures that we can efficiently schedule the tasks to maximize the total importance while avoiding overlaps.

<br><br>
The recurrence relation for the Weighted Interval Scheduling problem can be defined as follows:
<br><br>
Let OPT(i) represent the maximum total importance achievable by scheduling tasks up to and including task i.<br><br>
If we consider task i in our schedule, we need to find the last task (task j) that doesn't overlap with task i and has the maximum end time less than or equal to the start time of task i.<br><br>
We can then define the recurrence relation as:<br><br>
OPT(i) = max { Weight(i) + OPT(j), OPT(i - 1) }<br><br>
Here, Weight(i) represents the importance of task i.
OPT(j) represents the maximum total importance achievable up to the task that doesn't overlap with task i.

<h4>Q4. Mysterious Sorting (15 points)</h4>

<h5> Problem Statement:</h5>
Alice is trying to solve a tricky sorting problem. She is given a list of integers, and her task is to sort them using a mysterious sorting algorithm she has just discovered. The algorithm works as follows:
<br><br>
If the list contains only one element, it is already sorted.<br>
Otherwise, divide the list into two equal halves.<br>
Sort the left half using the mysterious sorting algorithm.<br>
Sort the right half using the mysterious sorting algorithm.<br>
Merge the two sorted halves into one sorted list.<br>
Alice is curious about the time complexity of her mysterious sorting algorithm. She believes it can be described using the Master Theorem for Divide and Conquer Recurrences. Can you help her analyze it?
<br><br>
Write a function mysterious_sort(arr) that takes an unsorted list of integers as input and returns the sorted list using Alice's mysterious sorting algorithm. Then, analyze the time complexity of her algorithm using the Master Theorem. Your analysis should include identifying the values of a, b, and f(n).
<br><br>
<strong>Input Format:</strong>
<br><br>
An unsorted list of integers, arr, where 1 <= n <= 10^6.
<br><br>
<strong>Output Format:</strong>
<br><br>
The sorted list obtained using the mysterious sorting algorithm.<br>
An analysis of the time complexity using the Master Theorem.
<br><br>
<strong>Sample Input:</strong>
<br>
<br>
arr = [12, 5, 8, 23, 14, 7, 18]<br>
sorted_arr, time_complexity = mysterious_sort(arr)
<br><br>
<strong>Sample Output:</strong>
<br><br>
Sorted List: [5, 7, 8, 12, 14, 18, 23]<br>
Time Complexity Analysis: O(n log n)
<br><br>
<strong>Constraints:</strong>
<br>
1. The input list arr contains integers.<br>
2. 1 <= n (the number of elements in arr) <= 10^6.<br>
3. The integers in arr can be positive or negative and are within the range of a 32-bit signed integer.<br>
4. The input list may contain duplicate values.<br>
5. The sorting algorithm should be implemented efficiently within the specified constraints.<br>
6. The time complexity analysis should adhere to the Master Theorem for divide and conquer algorithms.
<br><br>
<strong>Solution/PseudoCode:</strong>

```python
def mysterious_sort(arr):
    # Base case: If the list has only one element, it is already sorted.
    if len(arr) <= 1:
        return arr

    # Divide the list into two equal halves.
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort the left and right halves.
    left_sorted = mysterious_sort(left_half)
    right_sorted = mysterious_sort(right_half)

    # Merge the sorted halves to create a single sorted list.
    sorted_arr = merge(left_sorted, right_sorted)
    
    return sorted_arr

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

def time_complexity_analysis(arr):
    n = len(arr)
    if n <= 1:
        return "O(1)"
    return f"O(n * log({n}))"

# Example usage:
arr = [12, 5, 8, 23, 14, 7, 18]
sorted_arr = mysterious_sort(arr)
time_complexity = time_complexity_analysis(arr)
print("Sorted List:", sorted_arr)
print("Time Complexity Analysis:", time_complexity)
```

<strong>Justification:</strong>
<br><br>
The solution addresses the "Mysterious Sorting" problem by employing a divide-and-conquer approach. It recursively splits the input list into smaller halves until it reaches individual elements, which are inherently sorted. Then, it merges these sorted sublists while ensuring the overall merged list remains sorted. This divide-and-conquer strategy leverages the Master Theorem to estimate the time complexity. The approach guarantees a sorted output while efficiently breaking down the problem into smaller, manageable segments. It exhibits a time complexity of O(n * log(n)), making it an effective method for sorting large datasets.
