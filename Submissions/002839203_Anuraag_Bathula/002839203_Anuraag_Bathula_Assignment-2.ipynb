Use Kruskal's algorithm to find a minimum spanning tree for the connected weighted graph below:



What is the Time Complexity of Kruskal's algorithm?

Solution:
Pseudocode for Kruskal's algorithm:
1) Create an empty list MST to store the MST edges
2) Create set S storing all edges in graph
3) Sort all edges in the graph in ascending order of weight
4) Repeat following steps for each edge 
    - Select edge with smallest weight
    - Check if adding edge to MST creates a cycle in MST
    - If the above condition is not met, add edge to MST
    - Else if it is met, discard edge
5) Continue till MST has n-1 edges





Number of Vertices in graph = n = 6
Number of Edges in graph = m = 8

List of edges in ascending order- [DF, BD, BC, DE, CE, AB, EF, AF]

Creating MST using Kruskal's algorithm-  
Add DF to MST  
Add DB to MST  
Add BC to MST  
DE creates a cycle in MST so do not add DE  
Add CE to MST  
Add AB to MST  
We have added 5 edges to MST (number of vertices = 6 so 6-1 = 5 ) so MST is complete  
  
MST-[DF, BD, BC, CE, AB]  

Justification:
Time complexity of Kruskal's Algorithm is m*log(n)

The step 3 to sort all edges in the graph is O(m*log(m)) and is the most time consuming step in the algorithm. Since the step with the largest time complexity defines the overall time complexity in an algorithm (when the steps are carried out one after the other), the overall time complexity of the algorithm can be considered as O(m*log(m)).

However, since n < m^2 (in the case of a fully connected graph), m*log(m) can also be considered <= m*log(v^2) which is = 2m*log(n) which = O(m*log(n))

Hence we can argue the worst case time complexity of Kruskal's algorithm is O(m*log(n))

Master theorem  
The master theorem provides a solution in asymptotic terms (using Big O notation) for recurrence relations of types that occur in the analysis of many divide and conquer algorithms.  
The master theorem bounds asymptotically recurrence relations of the following form:  
T(n) = aT(n/b) + f(n), where a>=1 and b>1,  
Then T(n) can be classified as one of four forms to determine asymptotic bounds as follows:  
A. n^(log(base b)(a)) is polynomially larger than f(n) then T(n) = Big Theta(n^(log(base b)(a)))  
B. f(n) and n^(log(base b)(a))*log^k(n) are same size then T(n) = Big Theta(n^(log(base b)(a))*log^(k+1)(n))  
C. f(n) is polynomially larger than n^(log(base b)(a)) then T(n) = Big Theta(f(n))  
D. The master theorem does not apply.

Given the six intervals below, and their associated values; select a subset of non-overlapping intervals
with the maximum combined value. Use dynamic programming. 



Intervals = {A,B,C,D,E,F}  
Values "val" = {4,2,5,6,3,4}  
Previous Intervals "prev" = {null,null,null,A,C,C}  

Pseudocode-  
func(i){  
    if i<= 0 return 0  
    return max(val(i) + func(prev(i)), func(i-1))  
}  

|Interval|      Max|  
|--------|--------|
|A|             max(4+0,0) = 4|  
|B|             max(2+0,4) = 4|  
|C|             max(5+0,4) = 5|  
|D|             max(6+4,5) = 10|  
|E|             max(3+5,10) = 10|  
|F|             max(4+5,10) = 10|  

Finding maximum weighted interval  
|Interval|      Trace|                                         Final Set of Intervals  |
|--------|-----------|-----------------|
|F|             4+5<10|                                        {}  |
|E|             3+5<10 |                                       {}  |
|D|             6+4=10, add D to set and jump to prev(D)=A |   {D}  |
|C|                                                    |       {D}  |
|B|                                                    |       {D}  |
|A|             4+0=4, add A to set and jump to prev(A)=null|  {D,A}  |

Final set of intervals = {A,D} and total value = val(A) + val(D) = 4+6=10  

```python
Intervals = ["A","B","C","D","E","F"]
n = len(Intervals)
val = [4,2,5,6,3,4]
prev = [None,None,None,0,2,2]

arr = [0]*(n)

def func(i):
    if i<= 0: return 0
    if(prev[i]==None): p = 0
    else: p = prev[i]
    arr[i] = max(val[i] + func(p), func(i-1))
    return arr[i]

```

Given the weights and values of the five items in the table below, select a subset of items with the
maximum combined value that will fit in a knapsack with a weight limit, W, of 15. Use dynamic
programming.  

Number of items = n = 5  
Total weight of knapsack = W = 15
Items = {1,2,3,4,5}  
Weights = w = {3,4,6,2,5}  
Values = v ={6,10,14,8,10}  
Create empty array sized n*W

Pseudocode-  
for i from 1 to n  
    for x from i to W  
        if w[i] > x  
            arr[i][x] = arr[i-1][x]  
        else  
            arr[i][x] = max(arr[i-1][x], arr[i-1][x-w[i]] + v[i])

|---|---|---|---|---|---|
|15|6|16|30|38|38|
|14|6|16|30|32|34|

|3|6|6|6|8|8|
|2|0|0|0|8|8|
|1|0|0|0|0|0|

Tracing items through arr  
|Item and Weight remaining|      Trace|                                         Final Set of Items  |
|--------|-----------|-----------------|
|5,15|    arr[5][15] = 38 = arr[5][15] so this item isn't added to knapsack and we skip to item i-1=4   |{}  |
|4,15| arr[4][15] = 38 != arr[3][15] so this item is added to knapsack and we skip to i-1=3 and W-w[i]=13| {Item 4}|
|3,13| arr[3][13] = 30 != arr[2][13] so this item is added to knapsack and we skip to i-1=2 and W-w[i]=7| {Item 3,4}|
|2,7| arr[2][7] = 16 != arr[1][7] so this item is added to knapsack and we skip to i-1=1 and W-w[i]=3| {Item 2,3,4}|

Final list of items to be taken in knapsack = {Item 1, Item 2, Item 3, Item 4}  


```python

W = 15
Items = [1,2,3,4,5]
n = len(Items)
w = [3,4,6,2,5]
v = [6,10,14,8,10]

arr = [[0]*(W+1) for _ in range(n)]

for i in range(n):
    for x in range(W+1):
        if w[i] > x:
            arr[i][x] = arr[i-1][x]
        else:
            arr[i][x] = max(arr[i-1][x], arr[i-1][x-w[i]] + v[i])

print("Max capacity of knapsack is ",arr[n-1][W])
```

Q6

Calculate the Average value of levels in the following non-empty binary tree using an algorithm. What common graph searching algorithm did you use here and why?

                3
               / \
              7   10
             / \    \
            2   8    20

Solution:
A breadth first search approach fits this algorithm best because if run on the root node of the binary tree, it automatically guarantees that elements of each level are visited in sequence, which is not true for a depth first search approach.


```python
class Node():
    val = None
    left = None
    right = None
    def __init__(self, val, left, right):
        self.val = val
        self.left = left
        self.right = right

root = Node(3,Node(7,Node(2,None,None),Node(8,None,None)),Node(10,None,Node(20,None,None)))

ans = []
list = [root]
while(len(list)!=0):
    size = len(list)
    sum = 0
    count = 0
    for i in range(size):
        sum += list[0].val
        count += 1
        if list[0].left != None: list.append(list[0].left)
        if list[0].right != None: list.append(list[0].right)
        list.pop(0)
    ans.append(sum/count)

print(ans)
```

This modified breadth first search algorithm keeps tabs on sum of values and number of elements at the level and divides them to find the average value.

As we can see upon running the code, the averages are correctly found - [3.0,8.5,10.0]

Create two functions to calculate the nth Fibonacci number. Use a brute force approach in one function and use a dynamic programming approach in the other. Calculate the difference taken in time for various values of n

Solution:
We will create two functions bruteForce(n) and dynamicProg(n) that use brute force and dynamic programming respectively. We know that calculating the nth Fibonacci number without memoizing previous solutions is O(2^n) whereas calculating the nth Fibonacci number using dynamic programming is O(n), which is much better.

As we can see, the bruteforce algorithm takes 0.6 seconds to find the 35th Fibonacci number while the dynamicProg algorithm only takes 5.7*e^-5 seconds. It gets even worse when finding the 40th Fibonacci number where the bruteforce approach takes 6.4 seconds which is 10 times longer than previously whereas the dynamicprog approach takes the same 5*e^-5 seconds


```python
import time

def bruteForce(n):
    if n <= 1:
        return 0
    if n == 2:
        return 1
    return bruteForce(n-1) + bruteForce(n-2)

memo = []

def dynamicProg(n):
    if n <= 1:
        return 0
    if n == 2:
        return 1
    if memo[n] != -1:
        return memo[n]
    memo[n] = dynamicProg(n-1) + dynamicProg(n-2)
    return memo[n]


n = 35
memo = [-1]*(n+1)
start_time = time.time()
print(str(n) , "th Fibonacci number is " , str(bruteForce(n)) , "and time taken for algorithm using bruteForce() is " , (time.time() - start_time) ," seconds")

start_time = time.time()
print(str(n) , "th Fibonacci number is " , str(dynamicProg(n)) , "and time taken for algorithm using dynamicProg() is " , (time.time() - start_time) ," seconds")

n = 40
memo = [-1]*(n+1)
start_time = time.time()
print(str(n) , "th Fibonacci number is " , str(bruteForce(n)) , "and time taken for algorithm using bruteForce() is " , (time.time() - start_time) ," seconds")

start_time = time.time()
print(str(n) , "th Fibonacci number is " , str(dynamicProg(n)) , "and time taken for algorithm using dynamicProg() is " , (time.time() - start_time) ," seconds")
```
