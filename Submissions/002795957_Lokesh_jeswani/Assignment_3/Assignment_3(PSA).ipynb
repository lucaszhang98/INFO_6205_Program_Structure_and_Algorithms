Give a brief definition for the following:\
i. Depth-First Search (DFS)\
ii. Topological Sorting\
iii. Dijkstra's Algorithm\
iv. Minimum Spanning Tree (MST)\
v. Breadth-First Search (BFS)

<u>**Answer & Justification :**</u>\
i. Depth-First Search (DFS)
Depth-First Search is a graph traversal algorithm that explores as far as possible along a branch before backtracking. It is often used to explore or search through a graph or tree structure.

ii. Topological Sorting
Topological Sorting is a linear ordering of the vertices in a directed acyclic graph (DAG) such that for every directed edge (u, v), vertex u comes before vertex v in the ordering. It is commonly used in scheduling and dependency resolution.

iii. Dijkstra's Algorithm
Dijkstra's Algorithm is a graph algorithm that finds the shortest path from a source vertex to all other vertices in a weighted, non-negative edge graph. It's often used for finding optimal routes in networks or transportation systems.

iv. Minimum Spanning Tree (MST)
A Minimum Spanning Tree is a subgraph of a connected, undirected graph that includes all the vertices and a subset of the edges with the minimum possible total edge weight. MSTs are used in network design and clustering applications.

v. Breadth-First Search (BFS)
Breadth-First Search is a graph traversal algorithm that explores all the vertices at the current level before moving to the next level. It's often used to find the shortest path in unweighted graphs and is also essential in algorithms like Dijkstra's for weighted graphs.

<u><b>Q7.</b></u>

You are a hiker preparing for a journey in the wilderness, and you have a backpack with a limited weight capacity. You have a list of items, each with a specific weight and a value. Your goal is to maximize the total value of the items you can carry in your backpack without exceeding its weight limit.

Given the following items and their respective weights and values:

Item 1: Weight - 3 kg, Value - $50
Item 2: Weight - 5 kg, Value - $60
Item 3: Weight - 2 kg, Value - $40
Item 4: Weight - 6 kg, Value - $75
Item 5: Weight - 9 kg, Value - $100
Determine the optimal selection of items to maximize the total value within a backpack with a weight capacity of 10 kg.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the item weights and values
weights = [3, 5, 2, 6, 9]
values = [50, 60, 40, 75, 100]
capacity = 10
n = len(weights)

# Initialize the dynamic programming table
dp = np.zeros((n + 1, capacity + 1), dtype=int)

# Fill in the dynamic programming table
for i in range(1, n + 1):
    for w in range(1, capacity + 1):
        if weights[i - 1] <= w:
            dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
        else:
            dp[i][w] = dp[i - 1][w]

# Backtrack to find the selected items
selected_items = []
i, w = n, capacity
while i > 0 and w > 0:
    if dp[i][w] != dp[i - 1][w]:
        selected_items.append(i)
        w -= weights[i - 1]
    i -= 1

# Print the selected items and total value
print("Selected items:", selected_items)
print("Total value:", dp[n][capacity])

# Create a visualization of the dynamic programming table
plt.imshow(dp, cmap="Blues")
plt.colorbar()
plt.xlabel("Capacity (kg)")
plt.ylabel("Items")
plt.title("0/1 Knapsack Dynamic Programming Table")
plt.xticks(np.arange(capacity + 1))
plt.yticks(np.arange(n + 1), labels=["0"] + [f"Item {i}" for i in range(1, n + 1)])
plt.gca().invert_yaxis()
plt.show()
```

<u>**Answer & Justification :**</u>

To solve this problem, we can use the 0/1 Knapsack algorithm, a dynamic programming approach.

<b>Algorithm Steps:</b>

1. Create a table, "Knapsack," with rows representing the items (from 1 to 5) and columns representing the weights (from 0 to 10 kg).

2. Initialize the first row and the first column of the table to 0 since the maximum value with a weight of 0 kg or with no items is 0.

3. Fill in the table by considering two cases for each cell:

    - Case 1: If the weight of the current item (row) is greater than the current weight (column), copy the value from the cell above (the same weight without this item).
    - Case 2: If the weight of the current item is less than or equal to the current weight, consider taking the maximum value of two options:
        - Option A: The value from the cell above (the same weight without this item).
        - Option B: The value of the current item plus the value from the cell above and to the left (the remaining weight and the value if this item is added).
4. The value in the bottom-right cell of the table represents the maximum value that can be obtained with the given weight capacity.

5. To find the selected items, backtrack through the table, starting from the bottom-right cell. If the value in the cell is not the same as the value in the cell above, include the item, and move to the cell diagonally up and to the left. Repeat this process until you reach the top row.

<b><u>Justification:</u></b>

The 0/1 Knapsack algorithm maximizes the total value of items within a constrained weight capacity. By considering two cases for each item (either including it or excluding it), the algorithm determines the optimal selection of items to maximize the total value. The dynamic programming approach ensures an efficient solution. In this specific scenario, the optimal selection of items and their total value within a 10 kg weight limit can be found using this approach.

Junction B --(2)--> Junction D\
Junction C --(6)--> Junction D\

Junction D --(5)--> Junction E

Find the minimum cut-set in the water distribution network, which, if damaged or closed, would isolate one or more districts from the water supply while minimizing the total capacity of the cut pipes.

4. Calculate the total capacity of the minimum cut-set, which is the sum of the capacities of the saturated pipes.

Algorithm Steps:

1. Initialize the distance array, distance[], and the predecessor array, predecessor[]. For each node in the graph, set the distance to infinity (or a very large value) except for the source node, which is set to 0. Set the predecessor of each node to null to track the path.

2. Relax the edges repeatedly for a total of N-1 iterations, where N is the number of nodes in the graph. In each iteration, consider all edges and update the distance and predecessor arrays if a shorter path is found. The relaxation step involves comparing the distance from the source node to the current node plus the edge weight to the distance of the target node. If a shorter path is found, update the distance and predecessor arrays.

3. After N-1 iterations, check for the presence of negative-weight cycles. To do this, go through all the edges again. If any distance can be further minimized in this step, it indicates the existence of a negative-weight cycle in the graph. In this case, you can report their existence and stop the algorithm.

4. If no negative-weight cycles are found, the distance[] array will contain the shortest distances from the source node to all other nodes. The predecessor[] array can be used to trace back the shortest path from the source to any destination.

Using dynamic programming, demonstrate the process of selecting a subset of non-overlapping intervals with the highest total value from the following five intervals, each associated with specific values. Show your step-by-step solution.

Intervals and Values:

Interval 1: [1, 4], Value 5\
Interval 2: [3, 5], Value 2\
Interval 3: [6, 8], Value 7\
Interval 4: [8, 10], Value 2\
Interval 5: [9, 12], Value 8

<u>**Answer & Justification :**</u>

- Dynamic Programming Solution

```python
# Define the intervals and their values
intervals = [(1, 4), (3, 5), (6, 8), (8, 10), (9, 12)]
values = [5, 2, 7, 2, 8]

# Sort the intervals by their ending points
sorted_intervals = sorted(zip(intervals, values), key=lambda x: x[0][1])

# Initialize an array to store the maximum values for each interval
max_values = [0] * len(intervals)

# Initialize an array to store the selected intervals
selected_intervals = [None] * len(intervals)

# Dynamic programming approach to find the maximum combined value
for i in range(len(intervals)):
    max_values[i] = values[i]  # Initialize with the value of the interval
    selected_intervals[i] = [i]

    for j in range(i):
        if intervals[i][0] >= intervals[j][1]:
            if values[i] + max_values[j] > max_values[i]:
                max_values[i] = values[i] + max_values[j]
                selected_intervals[i] = selected_intervals[j] + [i]

# Find the interval with the maximum combined value
max_value_index = max(range(len(max_values)), key=lambda i: max_values[i])
selected_interval_indices = selected_intervals[max_value_index]

# Extract the selected intervals and their values
selected_intervals = [intervals[i] for i in selected_interval_indices]
selected_values = [values[i] for i in selected_interval_indices]

# Calculate the maximum combined value
maximum_combined_value = max_values[max_value_index]

print("Selected Intervals:", selected_intervals)
print("Selected Values:", selected_values)
print("Maximum Combined Value:", maximum_combined_value)

```

<u><b>Justification:</b></u>

This dynamic programming solution first sorts the intervals by their ending points to ensure that non-overlapping intervals are processed in the correct order. It then iterates through the intervals and uses dynamic programming to find the maximum combined value for each interval. The algorithm considers whether adding the current interval to a previous non-overlapping set of intervals would yield a higher value.

In this example, the selected intervals with the maximum combined value are [1, 4], [6, 8], and [9, 12], with a maximum combined value of 20. This approach ensures that you maximize the total value while selecting non-overlapping intervals.
