* Dijkstra's Algorithm:

Dijkstra's Algorithm is a popular and efficient algorithm for finding the shortest path from a source node to all other nodes in a non-negative weighted graph. It works by iteratively selecting the vertex with the smallest known distance from the source and gradually building the shortest paths.

Example: Finding the shortest path from node A to all other nodes in a road network with distances between intersections.

* Pseudocode:

function Dijkstra(graph, source):
    initialize distances and predecessors
    create a priority queue Q

    for each vertex v in graph:
        if v is the source, set distance to 0, else set it to infinity
        add v to Q

    while Q is not empty:
        u = vertex in Q with the smallest distance
        remove u from Q

        for each neighbor v of u:
            alt = distance[u] + weight(u, v)
            if alt < distance[v]:
                distance[v] = alt
                set predecessor of v to u

    return distances and predecessors

* Graph Traversal:

Graph Traversal is a fundamental operation in graph theory and computer science. It involves systematically visiting all vertices and edges of a graph. Two common graph traversal algorithms are:
**Depth-First Search (DFS):** DFS explores as far as possible along each branch before backtracking. It's often used to find connected components and detect cycles.
**Breadth-First Search (BFS):** BFS explores all the vertices at the current depth before moving to the next depth. It's often used to find the shortest path.

Example: Navigating a maze using DFS to explore paths deeply or using BFS to find the shortest path.

* Pseudocode (DFS):

function DFS(graph, start):
    create a stack S
    push start onto S
    mark start as visited

    while S is not empty:
        current = pop S
        for each neighbor v of current:
            if v is not unvisited:
                push v onto S
                mark v as visited

* Pseudocode (BFS):

function BFS(graph, start):
    create a queue Q
    enqueue start into Q
    mark start as visited

    while Q is not empty:
        current = dequeue Q
        for each neighbor v of current:
            if v is not visited:
                enqueue v into Q
                mark v as visited

* Dynamic Programming:

Dynamic Programming is a technique used to solve complex problems by breaking them down into simpler overlapping subproblems. It stores the results of each subproblem to avoid redundant calculations. Dynamic programming is particularly useful for optimization and search problems.

Example: The Fibonacci sequence calculation, which can be optimized using dynamic programming (Memoization or Tabulation).

* Pseudocode:

function Fibonacci(n):
    create a table fib of size n+1
    set fib[0] to 0 and fib[1] to 1

    for i from 2 to n:
        fib[i] = fib[i-1] + fib[i-2]

    return fib[n]

* Greedy Algorithm:

A Greedy Algorithm is an approach to problem-solving that makes the best choice at each step based on the current information, without considering the global picture. Greedy algorithms are used when a series of decisions must be made, each influencing the next. They aim to find a globally optimal solution through a sequence of locally optimal choices.

Example: The classic "Coin Change" problem, where you aim to make change using the fewest coins by selecting the largest coin at each step.

* Pseudocode:

function GreedyCoinChange(coins, amount):
    sort coins in decreasing order
    count = 0
    i = 0

    while amount > 0:
        if coins[i] <= amount:
            amount -= coins[i]
            count++
        else:
            i++

    return count

* Topological Sorting:

Topological Sorting is a specific ordering of the vertices in a directed acyclic graph (DAG) such that, for every directed edge (u, v), vertex u comes before vertex v in the ordering. It's often used to represent dependencies or sequencing in various applications, such as scheduling tasks.

Example: Organizing tasks with dependencies in a project to ensure they are executed in the correct order.

* Pseudocode:

function TopologicalSort(graph):
    create an empty list result
    create a stack S
    mark all vertices as unvisited

    for each vertex v in graph:
        if v is unvisited:
            Visit(v, S)

    return result

function Visit(vertex, S):
    mark vertex as visited
    for each neighbor w of vertex:
        if w is unvisited:

```python
import networkx as nx

def shortest_path_for_package_delivery(graph, packages, central_hub):
    # Step 1: Initialize distances
    distances = nx.single_source_bellman_ford(graph, central_hub)[0]

    # Step 2: Create variables
    optimal_delivery_order = []
    total_delivery_time = 0

    # Step 3: Find the Optimal Delivery Order
    packages.sort(key=lambda package: distances[package['destination']])

    # Step 4: Report the Results
    for package in packages:
        destination = package['destination']
        time_limit = package['time_limit']

        if distances[destination] <= time_limit:
            optimal_delivery_order.append(destination)
            total_delivery_time += distances[destination]
        else:
            print(f"Package to {destination} is undeliverable within the time limit.")

    return optimal_delivery_order, total_delivery_time

# Example usage
G = nx.DiGraph()
G.add_edge('Hub', 'A', weight=5)
G.add_edge('Hub', 'B', weight=8)
G.add_edge('A', 'C', weight=6)
G.add_edge('B', 'C', weight=7)
G.add_edge('C', 'Hub', weight=10)

packages = [
    {'destination': 'A', 'time_limit': 8},
    {'destination': 'B', 'time_limit': 9},
    {'destination': 'C', 'time_limit': 12}
]
central_hub = 'Hub'

optimal_order, total_time = shortest_path_for_package_delivery(G, packages, central_hub)
print("Optimal Delivery Order:", optimal_order)
print("Total Delivery Time:", total_time)
```

# Solution:

The Ford-Fulkerson algorithm is a widely-used method for finding the maximum flow in a flow network. A flow network is a directed graph in which each edge is associated with a capacity, indicating the maximum amount of flow it can carry. The goal of the algorithm is to determine the maximum flow from a designated source node to a sink node while respecting the capacity constraints of the edges.

The algorithm operates by iteratively finding augmenting paths from the source to the sink in a residual graph. A residual graph is a modified representation of the original network that reflects the remaining capacity on each edge. The algorithm identifies the bottleneck, which is the minimum capacity along the augmenting path, and increases the flow through these edges. This process continues until no more augmenting paths can be found in the residual graph.

The Ford-Fulkerson algorithm can provide not only the maximum flow but also the flow distribution on each edge in the network, making it a valuable tool for optimizing the transportation of resources in various real-world scenarios, such as network flow in transportation, communication, and supply chain management.

* To solve the "Ford-Fulkerson Maximum Flow" problem, we can use the Ford-Fulkerson algorithm. Here's a step-by-step solution:

**Step 1:** Initialize the flow on each pipe to 0, and create a residual graph with the same capacities as the original graph.

**Step 2:** While there is a path from the source (A) to the target (E) in the residual graph:

   a. Find the minimum capacity (bottleneck capacity) along this path.

   b. Increase the flow on each pipe in the path by the bottleneck capacity.

   c. Decrease the capacity of forward edges by the bottleneck capacity, and increase the capacity of backward edges (back edges) by the same amount.

**Step 3:** The maximum flow is the sum of flows leaving the source node (A), and the flow on each pipe is determined after the algorithm completes.

# Pseudocode:

function FordFulkerson(graph, source, sink):
    Initialize flow on all edges to 0

    while there is an augmenting path P from source to sink in the residual graph:
        Find the bottleneck capacity (min capacity) on path P
        Augment the flow on path P by the bottleneck capacity
        Update the residual graph by decreasing capacity on forward edges and increasing capacity on backward edges

    Compute the maximum flow by summing the flow leaving the source node

    return the maximum flow


```python
import networkx as nx

def ford_fulkerson_max_flow(graph, source, sink):
    # Initialize the flow on all edges to 0
    flow = {(u, v): 0 for u, v in graph.edges}
    
    def find_augmenting_path(graph, source, sink, flow):
        # Implement a path-finding algorithm to find an augmenting path
        visited = set()
        queue = [(source, [source])]
        
        while queue:
            node, path = queue.pop(0)
            for neighbor in graph.neighbors(node):
                if neighbor not in visited and (node, neighbor) in flow:
                    new_path = list(path)
                    new_path.append(neighbor)
                    new_edge = (node, neighbor)
                    if flow[new_edge] < graph.edges[node, neighbor]['capacity']:
                        if neighbor == sink:
                            return new_path
                        queue.append((neighbor, new_path))
                        visited.add(neighbor)
        
        return None
    
    path = find_augmenting_path(graph, source, sink, flow)
    while path:
        # Find the bottleneck capacity on the path
        bottleneck = min(graph.edges[node, neighbor]['capacity'] - flow[(node, neighbor)] for node, neighbor in zip(path[:-1], path[1:]))
        
        # Augment the flow on the path by the bottleneck capacity
        for node, neighbor in zip(path[:-1], path[1:]):
            edge = (node, neighbor)
            flow[edge] += bottleneck
            reverse_edge = (neighbor, node)
            if reverse_edge not in flow:
                flow[reverse_edge] = 0
        
        path = find_augmenting_path(graph, source, sink, flow)
    
    # Compute the maximum flow by summing the flow leaving the source node
    max_flow = sum(flow[(source, v)] for v in graph.neighbors(source))
    
    return max_flow, flow

# Create a directed graph with capacities and initial flows
G = nx.DiGraph()
G.add_edge('A', 'E', capacity=20)
G.add_edge('A', 'B', capacity=15)
G.add_edge('A', 'C', capacity=12)
G.add_edge('B', 'C', capacity=18)
G.add_edge('B', 'D', capacity=10)
G.add_edge('C', 'E', capacity=15)
G.add_edge('C', 'D', capacity=20)
G.add_edge('D', 'E', capacity=15)

# Set initial flows
initial_flows = {
    ('A', 'E'): 10,
    ('A', 'B'): 8,
    ('A', 'C'): 12,
    ('B', 'C'): 10,
    ('B', 'D'): 10,
    ('C', 'E'): 5,
    ('C', 'D'): 15,
    ('D', 'E'): 6
}

for edge, initial_flow in initial_flows.items():
    G.edges[edge]['initial_flow'] = initial_flow

source_node = 'A'
sink_node = 'E'

max_flow, final_flow = ford_fulkerson_max_flow(G, source_node, sink_node)
print("Maximum Flow:", max_flow)
print("Final Flow on Each Edge:")
for edge, flow in final_flow.items():
    print(f"{edge}: {flow}")
```

# Solution:

To find the minimum cost flow in the road network while respecting capacity constraints and the limit on the number of available trucks, we can use the Minimum Cost Flow algorithm. The idea is to modify the capacities and costs of the edges based on the current flow and iteratively update the flow to minimize the total cost.

* Here is a step-by-step solution to the problem:

Initialize the flow on each edge to zero.

While there are available trucks:
a. Build a residual graph with updated capacities and costs based on the current flow.
b. Use a shortest path algorithm (such as Dijkstra's) to find the shortest path from the factory to the warehouse in the residual graph. The cost of the path should be based on the modified edge costs.
c. If no path is found, exit the loop.
d. Determine the maximum flow that can be sent along the path without exceeding the available trucks or edge capacities.
e. Update the flow and edge costs based on the determined flow value.

Calculate the total cost of transportation, which is the sum of costs on each edge in the final flow.

Output the minimum cost flow and the total cost of transportation.

This algorithm ensures that the flow respects capacity constraints and the limit on the number of trucks while minimizing the total cost of transportation.

The time complexity of this algorithm depends on the shortest path algorithm used and the number of iterations. In practice, it runs efficiently on road networks with reasonable sizes.

Network Construction: Constructing the flow network from the road network takes O(E) time, where E is the number of road segments.

Network Simplex Algorithm: The most time-consuming part is solving the minimum cost flow problem using the Network Simplex algorithm. The worst-case complexity of the Network Simplex algorithm is O(V^3), where V is the number of nodes in the flow network.

In practice, the Network Simplex algorithm often runs in nearly linear time on real-world instances, making it efficient.
Overall Complexity: The overall complexity of the Minimum Cost Flow algorithm is determined by the time required for network construction and solving the MCF problem.

```python
import matplotlib.pyplot as plt

def plot_gantt_chart(tasks, optimal_subset):
    fig, ax = plt.subplots(figsize=(10, 2))

    for i, (start, end, value) in enumerate(tasks):
        color = "g" if i in optimal_subset else "r"
        ax.barh(0, end - start, left=start, height=0.2, color=color, label=f'Task {i + 1}\nValue: {value}')
    
    ax.set_xlabel('Time')
    ax.set_yticks([0])
    ax.set_yticklabels(['Tasks'])
    ax.legend(loc='upper left')
    plt.show()

# Example usage:
tasks = [(0, 3, 5), (1, 4, 7), (2, 6, 9), (5, 7, 4)]
optimal_subset = [0, 3]
plot_gantt_chart(tasks, optimal_subset)
```

# Solution:

The problem at hand can be seen as a variation of the well-known "Weighted Interval Scheduling" problem, which is a classic optimization problem.

* Algorithm:
Sort the tasks by their end times in ascending order. Sorting the tasks is a crucial step because it allows us to consider tasks in a specific order. By sorting them based on their end times, we ensure that, when choosing a task, it doesn't conflict with tasks that end later.

Initialize two arrays:

dp, a dynamic programming array to store the maximum value achievable up to each task.
prev, an array to store the index of the previous task contributing to the maximum value.
Dynamic Programming Formulation:

For each task i in the sorted order:

Find the maximum value dp[i] that can be achieved by either selecting or excluding the task.
If the task is selected, its value is added, and we consider the last non-overlapping task before it.
If the task is excluded, we simply copy the value from the previous task.
The recurrence relation can be written as:
dp[i] = max(dp[i-1], dp[last_non_overlapping_task] + value[i])

Where last_non_overlapping_task is the index of the last task that doesn't overlap with task i.

Reconstructing the Optimal Subset:

We can reconstruct the optimal subset by starting from the task with the maximum value and tracing back to the first task.

* Time Complexity:
The time complexity of this algorithm is O(n log n) due to the initial sorting of tasks, where n is the number of tasks. The subsequent dynamic programming step has a time complexity of O(n).

* Space Complexity:
The space complexity is O(n) to store the dynamic programming array and prev array.


```python
def max_combined_value(tasks):
    # Sort tasks by their end times
    tasks.sort(key=lambda x: x[2])

    n = len(tasks)
    dp = [0] * (n + 1)

    for i in range(1, n + 1):
        # Find the index of the last non-overlapping task
        j = binary_search(tasks, i)
        dp[i] = max(dp[i - 1], tasks[i - 1][1] + dp[j])

    return dp[n]

def binary_search(tasks, current_task):
    low, high = 0, current_task - 1
    while low <= high:
        mid = (low + high) // 2
        if tasks[mid][2] <= tasks[current_task - 1][1]:
            if tasks[mid + 1][2] <= tasks[current_task - 1][1]:
                low = mid + 1
            else:
                return mid
        else:
            high = mid - 1
    return -1

# Example usage:
tasks = [(0, 3, 5), (1, 4, 7), (2, 6, 9), (5, 7, 4)]
result = max_combined_value(tasks)
print("Maximum Combined Value:", result)
```

   - Task 2 contributes 9 to the total value (5 from task 0 and 4 from task 2).

   The optimal subset is [(0, 3, 5), (2, 6, 9)] with a total value of 13.

# Solution:

Step 1: Understanding the Problem
The problem involves selecting a subset of items to maximize the total pleasure while staying within the weight limit of the backpack. Each item has a weight and a pleasure score, and the goal is to maximize the total pleasure without exceeding the maximum weight capacity.

Step 2: Dynamic Programming Approach
To solve this problem efficiently, we can use dynamic programming. We'll create a 2D table to store the maximum pleasure that can be achieved at each combination of items and backpack weight.

Step 3: Create a Table
Create a 2D table, dp, where dp[i][j] represents the maximum pleasure that can be achieved using the first i items with a maximum backpack weight of j. Initialize all values in the table to 0.

Step 4: Filling the Table
We'll fill in the dp table using a dynamic programming approach. Start by iterating through each item and each possible backpack weight.

For each item i (from 1 to the total number of items), iterate through the possible backpack weights j (from 0 to the maximum weight capacity W).

For each combination of item i and backpack weight j, calculate two values:

Value 1: The maximum pleasure achieved by including item i in the backpack. This is dp[i-1][j-w[i]] + p[i], where w[i] is the weight of item i and p[i] is the pleasure score of item i. Ensure that j-w[i] is non-negative.
Value 2: The maximum pleasure achieved by excluding item i. This is dp[i-1][j].
Update dp[i][j] as the maximum of Value 1 and Value 2. In other words, dp[i][j] = max(Value 1, Value 2).

Step 5: Retrieve the Solution
After filling the dp table, the maximum total pleasure can be found in dp[n][W], where n is the total number of items and W is the maximum weight capacity. This value represents the maximum total pleasure achievable without exceeding the weight limit.

Step 6: Trace Back for Selected Items
To determine which items were selected to achieve the maximum pleasure, you can trace back through the dp table. Start from dp[n][W] and move backward while selecting the items that contributed to the maximum pleasure. Track the items by comparing dp[i][j] with dp[i-1][j-w[i]] + p[i]. If the two values are not equal, it means that item i was selected.

Step 7: Complexity Analysis
The time complexity of this dynamic programming approach is O(n * W), where n is the number of items and W is the maximum weight capacity.

In summary, the "maximize_pleasure" problem can be efficiently solved using dynamic programming. It finds the subset of items that maximizes pleasure while respecting the weight limit of the backpack.

# Pseudocode:

function maximize_pleasure(items, W):
    n = length(items)  # Number of items
    dp = 2D array of size (n+1) x (W+1)  # Initialize a DP table

    # Initialize the DP table with zeros
    for i from 0 to n:
        dp[i][0] = 0  # If the weight capacity is 0, no pleasure can be achieved
    for j from 0 to W:
        dp[0][j] = 0  # If there are no items, no pleasure can be achieved

    # Fill the DP table
    for i from 1 to n:
        for j from 1 to W:
            w_i = weight of item i
            p_i = pleasure score of item i

            if w_i > j:
                # Item i cannot be included, so the maximum pleasure is the same as without it
                dp[i][j] = dp[i-1][j]
            else:
                # Calculate the maximum pleasure by either including or excluding item i
                include_i = dp[i-1][j - w_i] + p_i
                exclude_i = dp[i-1][j]
                dp[i][j] = max(include_i, exclude_i)

    # The maximum total pleasure is in dp[n][W]
    max_pleasure = dp[n][W]

    # Trace back to find selected items
    selected_items = []
    j = W
    for i from n down to 1:
        if dp[i][j] != dp[i-1][j]:
            selected_items.append(i)
            j = j - weight of item i

    return max_pleasure, selected_items


* Complexity Analysis:
Time Complexity: O(n * W), where n is the number of items and W is the maximum weight capacity. The nested loops to fill the DP table take O(n * W) time.

Space Complexity: O(n * W), as we use a 2D DP table of size (n+1) x (W+1).

This dynamic programming approach efficiently finds the maximum total pleasure and the selected items within the given weight limit.

```python
def maximize_pleasure(items, W):
    n = len(items)
    dp = [[0] * (W + 1) for _ in range(n + 1)]

    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif items[i - 1][0] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - items[i - 1][0]] + items[i - 1][1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][W]

# Example usage:
items = [(2, 5), (3, 7), (4, 6), (1, 3)]
W = 5
max_pleasure = maximize_pleasure(items, W)
print(max_pleasure)  # Output: 12
```

We'll define the recurrence relation for dp[i] as follows:
dp[i] = max(dp[i-1], dp[i-2] + poster_promotion[i] - poster_placement_cost, dp[i-1] + poster_promotion[i])

dp[i-1]: Represents the maximum promotional value obtained without placing a poster on block i.

dp[0] = poster_promotion[0] (if there's only one block, we place the poster on it)
dp[1] = max(poster_promotion[0], poster_promotion[1]) (we choose the block with a higher promotional value)
Now, let's outline the steps of the dynamic programming algorithm:

Initialize a list dp of length n to store the maximum promotional values for each block.
Set the base cases for dp[0] and dp[1].
Iterate from block i = 2 to n-1:
Use the recurrence relation to update dp[i].
The final answer is dp[n-1], which represents the maximum combined promotional value for all blocks.

The time complexity of this dynamic programming solution is O(n) because we iterate through all blocks once. The space complexity is O(n) to store the dp list.

```python
def maximize_poster_promotion(block_appeal, poster_promotion, poster_placement_cost):
    n = len(block_appeal)
    if n == 0:
        return 0
    if n == 1:
        return poster_promotion[0]

    dp = [0] * n
    dp[0] = poster_promotion[0]
    dp[1] = max(poster_promotion[0], poster_promotion[1])

    for i in range(2, n):
        dp[i] = max(dp[i-1], dp[i-2] + poster_promotion[i] - poster_placement_cost, dp[i-1] + poster_promotion[i])

    return dp[n-1]

# Example usage:
block_appeal = [3, 2, 4, 5, 1]
poster_promotion = [8, 6, 12, 10, 3]
poster_placement_cost = 2
max_promotion = maximize_poster_promotion(block_appeal, poster_promotion, poster_placement_cost)
print(max_promotion)  # Output: 33
```
