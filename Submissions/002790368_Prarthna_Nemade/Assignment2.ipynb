# Question-1: *Using Kruskal's algorithm to find a minimum spanning tree for the connected weighted graph.*

# Problem Statement:
Given a connected, undirected graph with weighted edges, the task is to implement Kruskal's algorithm to find the minimum spanning tree (MST) of this graph. consider the following graph:

Vertices: A, B, C, D, E
Edges and Weights:

AB with weight 2
AC with weight 4
AD with weight 1
BC with weight 3
BD with weight 2
BE with weight 5
CD with weight 4
CE with weight 1

* Input:

The number of vertices in the graph, n, is equal to the dimension of the graph matrix.

* Output:
Return a list of edges that form the minimum spanning tree (MST) of the input graph. Each edge should be represented as a tuple (u, v, weight), where u and v are the vertices connected by the edge, and weight is the weight of the edge.

* Constraints:

0 ≤ graph[i][j] ≤ 10^9 for all 0 ≤ i, j < n
The graph is connected and contains no self-loops.

# Solution:

Applying Kruskal's algorithm to find the minimum spanning tree step by step:

* Step 1: Initialization
Create a set of all edges and sort them by weight in ascending order.
Initialize an empty set for the minimum spanning tree (MST).

* Step 2: Adding Edges
Start adding edges from the sorted list of edges with the smallest weight first, while ensuring that adding each edge doesn't create a cycle in the MST.
Initially, add AD with weight 1 to the MST.
MST: {AD}

* Step 3: Continue Adding Edges
Next, add CE with weight 1 to the MST.
MST: {AD, CE}

* Step 4: Continue Adding Edges
Add BD with weight 2 to the MST.
MST: {AD, CE, BD}

* Step 5: Continue Adding Edges
Add AB with weight 2 to the MST.
MST: {AD, CE, BD, AB}

* Step 6: Continue Adding Edges
Add DE with weight 3 to the MST.
MST: {AD, CE, BD, AB, DE}

* Step 7: Continue Adding Edges
At this point, all vertices are included in the MST, and the algorithm terminates.
The minimum spanning tree for this graph is {AD, CE, BD, AB, DE}, and its total weight is 1 + 1 + 2 + 2 + 3 = 9.

* Pseudo-code:
def kruskal(graph):
    MST = []  # Initialize an empty list to store the minimum spanning tree
    edges = []  # Initialize a list to store all edges in the graph
    DSU = DisjointSetUnion(n)  # Initialize a disjoint-set data structure
    n = len(graph)  # Get the number of vertices in the graph

    # Step 1: Collecting all edges from the graph
    for i in range(n):
        for j in range(i + 1, n):
            if graph[i][j] > 0:
                edges.append((i, j, graph[i][j]))  # Store edge (u, v, weight)

    # Step 2: Sorting the edges in ascending order of their weights
    edges.sort(key=lambda x: x[2])

    # Step 3: Iterating through the sorted edges
    for edge in edges:
        u, v, weight = edge
        # Check if adding this edge forms a cycle in MST
        if DSU.find(u) != DSU.find(v):
            MST.append((u, v, weight))  # Add the edge to MST
            DSU.union(u, v)  # Union the sets containing u and v

    # Step 4: Return the minimum spanning tree
    return MST


```python
# code:

class DisjointSetUnion:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.rank[root_x] < self.rank[root_y]:
                self.parent[root_x] = root_y
            elif self.rank[root_x] > self.rank[root_y]:
                self.parent[root_y] = root_x
            else:
                self.parent[root_y] = root_x
                self.rank[root_x] += 1

def kruskal(graph):
    n = len(graph)
    edges = []
    for i in range(n):
        for j in range(i + 1, n):
            if graph[i][j] != 0:
                edges.append((i, j, graph[i][j]))
    
    edges.sort(key=lambda x: x[2])  # Sorting the edges by their weight
    mst = []
    dsu = DisjointSetUnion(n)
    
    for edge in edges:
        u, v, weight = edge
        if dsu.find(u) != dsu.find(v):
            mst.append(edge)
            dsu.union(u, v)
    
    return mst

graph = [
    [0, 2, 4, 1, 0],
    [2, 0, 3, 0, 0],
    [4, 3, 0, 0, 5],
    [1, 0, 0, 0, 6],
    [0, 0, 5, 6, 0]
]

minimum_spanning_tree = kruskal(graph)
for edge in minimum_spanning_tree:
    print(f"Edge: {edge[0]} - {edge[1]}, Weight: {edge[2]}")
```

# Question-2: *Another way to perform topological sorting on a directed acyclic graph G=(V, E) is to repeatedly find a vertex of in-degree 0, output it, and remove it and all of its outgoing edges from the graph. # Problem Statement:

# Solution:

* Topological Sorting:
Topological sorting is a linear ordering of the vertices (streets in our case) in a directed acyclic graph (DAG) such that for every directed edge (u, v), vertex u comes before vertex v in the ordering. It is a fundamental algorithm used in various scheduling and dependency-related problems.

The problem revolves around identifying "safe nodes" in a directed graph. A node is considered safe if there exists a path from that node leading only to terminal nodes (nodes with no outgoing edges). The goal is to find and return all such safe nodes sorted in ascending order.

* Theory:

We can solve this problem by utilizing a topological sorting approach. The key insight is that any node with an out-degree of 0 must be a safe node.

To achieve this, we start by creating a reverse graph, where edges are reversed compared to the original graph. This allows us to find incoming edges for each node.

We maintain an out_degree array to keep track of the out-degrees for each node in the original graph.

Initialize a queue with nodes having an out-degree of 0 since they are potential safe nodes.

Perform a topological sort on the reverse graph. While doing this, we decrease the out-degrees of the parent nodes whenever we process a child node.

As we visit nodes with an out-degree of 0, we add them to the result list of safe nodes.

Finally, we return the list of safe nodes sorted in ascending order.

# Pseudocode:

function eventualSafeNodes(graph):
    n = number of nodes in the graph
    out_degree = array of size n initialized with 0
    reverse_graph = create an empty graph with n nodes
    safe_nodes = empty list

    # Create the reverse graph and calculate out-degrees
    for i in range(n):
        for neighbor in graph[i]:
            Add edge from neighbor to i in reverse_graph
            Increment out_degree[i]

    queue = empty queue

    # Initialize the queue with nodes having out-degree 0
    for i in range(n):
        if out_degree[i] == 0:
            enqueue i to queue

    # Perform topological sort
    while queue is not empty:
        node = dequeue from queue
        Add node to safe_nodes
        for parent in reverse_graph[node]:
            Decrement out_degree[parent]
            if out_degree[parent] == 0:
                enqueue parent to queue

    Return safe_nodes sorted in ascending order

* Complexity Analysis:
Time Complexity: O(V + E), where V is the number of nodes and E is the number of edges in the graph. This is because we perform a topological sort, which takes linear time.

Space Complexity: O(V + E), where V is the space required to store the graph, and E is the space required for the out_degree array and the queue.

* Final Output:
The function eventualSafeNodes returns a list of safe nodes in ascending order.

```python
def eventualSafeNodes(graph):
    n = len(graph)
    out_degree = [0] * n
    reverse_graph = [[] for _ in range(n)]
    safe_nodes = []

    # Create the reverse graph and calculate out-degrees of nodes
    for i in range(n):
        for neighbor in graph[i]:
            reverse_graph[neighbor].append(i)
            out_degree[i] += 1

    # Initialize a queue with nodes having out-degree 0
    queue = []
    for i in range(n):
        if out_degree[i] == 0:
            queue.append(i)

    # Perform a topological sort
    while queue:
        node = queue.pop(0)
        safe_nodes.append(node)
        for parent in reverse_graph[node]:
            out_degree[parent] -= 1
            if out_degree[parent] == 0:
                queue.append(parent)

    return sorted(safe_nodes)

# Example usage:
graph1 = [[1,2],[2,3],[5],[0],[5],[],[]]
graph2 = [[1,2,3,4],[1,2],[3,4],[0,4],[]]

print(eventualSafeNodes(graph1))  # Output: [2, 4, 5, 6]
print(eventualSafeNodes(graph2))  # Output: [4]
```

This solution uses topological sorting to find all safe nodes. It first creates a reverse graph, calculates out-degrees of nodes, and initializes a queue with nodes having out-degree 0. Then, it performs a topological sort starting from nodes with out-degree 0, gradually reducing the out-degrees of their parents. Finally, it returns a sorted list of safe nodes.

# Question-3: *Use Prim's algorithm to find a minimum spanning tree in a graph of your choice. Show your steps.*

# Problem statement:
The task is to find a minimum spanning tree (MST) in a connected, weighted graph G=(V, E) using Prim's algorithm. The goal is to construct a subgraph that includes all vertices from the original graph while minimizing the total edge weight. Implement the Prim's algorithm and illustrate its steps on a provided graph.

# Input:
The graph is represented by a set of vertices V and a set of weighted edges E.
V is a list of vertices, where each vertex is represented by a unique identifier.
E is a list of weighted edges, where each edge is represented by a tuple (u, v, weight) indicating an edge from vertex u to vertex v with the given weight.
V = [A, B, C, D, E]
E = [(A, B, 4), (A, C, 1), (B, C, 2), (B, D, 5), (C, D, 8), (C, E, 10), (D, E, 2)]


# Output:
Return a list of edges that form the minimum spanning tree, represented as tuples (u, v, weight).
('A', 'C', 1)
('C', 'B', 2)
('B', 'D', 5)
('D', 'E', 2)

# Constraints:
The input graph G is connected.
The number of vertices V is in the range 2 ≤ |V| ≤ 10^4.
The number of edges E is in the range 1 ≤ |E| ≤ min(10^4, |V| * (|V| - 1) / 2).
Vertices are represented by unique identifiers.
Edge weights are positive integers.

# Solution:

* Steps to Find MST:

Start with an empty set that will eventually contain the MST.
Initialize a priority queue (min-heap) to store candidate edges with their weights.
Randomly select an initial vertex from V and add it to the MST.
For each edge incident to the selected vertex, add it to the priority queue.
While the priority queue is not empty:
Dequeue the edge with the smallest weight from the priority queue.
If adding this edge to the MST does not create a cycle, add it to the MST.
Otherwise, discard the edge.
Continue this process until the MST includes all vertices.

* Pseudocode:

Prim's Algorithm for Minimum Spanning Tree (MST):

Input: Graph G = (V, E) represented by vertices V and weighted edges E.

1. Initialize an empty set MST to store the MST.
2. Initialize a set visited to keep track of visited vertices (start with an empty set).
3. Select a starting vertex arbitrarily from V and add it to visited.
4. Initialize a priority queue (min-heap) Q with the edges incident to the selected vertex.

5. While Q is not empty:
     6. Dequeue the edge (u, v, weight) with the smallest weight from Q.
     7. If v is not in visited:
         8. Add v to visited.
         9. Add the edge (u, v, weight) to MST.
        10. Enqueue all edges incident to v that connect to unvisited vertices into Q.

11. Return MST.

Output: Minimum Spanning Tree MST represented as a set of edges.

This pseudocode outlines the steps for finding the Minimum Spanning Tree using Prim's algorithm. It starts with an empty MST, selects a starting vertex, and iteratively adds edges with the smallest weights to the MST while ensuring that no cycles are formed. The algorithm continues until all vertices are included in the MST.

```python
# code:

import heapq

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, weight):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))

    def prim_mst(self):
        mst = []          # Initialize MST
        visited = set()   # Set to keep track of visited vertices
        start_vertex = list(self.graph.keys())[0]  # Start from the first vertex

        # Priority queue (min-heap) to store candidate edges
        priority_queue = [(0, None, start_vertex)]

        while priority_queue:
            weight, parent, current_vertex = heapq.heappop(priority_queue)

            if current_vertex not in visited:
                visited.add(current_vertex)
                if parent is not None:
                    mst.append((parent, current_vertex, weight))

                for neighbor, edge_weight in self.graph[current_vertex]:
                    if neighbor not in visited:
                        heapq.heappush(priority_queue, (edge_weight, current_vertex, neighbor))

        return mst

# Example usage:
g = Graph()
g.add_edge('A', 'B', 4)
g.add_edge('A', 'C', 1)
g.add_edge('B', 'C', 2)
g.add_edge('B', 'D', 5)
g.add_edge('C', 'D', 8)
g.add_edge('C', 'E', 10)
g.add_edge('D', 'E', 2)

minimum_spanning_tree = g.prim_mst()
print("Minimum Spanning Tree Edges:")
for edge in minimum_spanning_tree:
    print(edge)
```

# Solution:

* Algorithm Explanation:

The algorithm used in this problem is based on topological sorting, a technique commonly applied to Directed Acyclic Graphs (DAGs). The goal is to find the optimal order of completing tasks to minimize the total project completion time while respecting task dependencies (prerequisites).

* Steps to the algorithm:

Creating a Task Map: We start by visualizing our tasks and their dependencies as a directed graph. Think of each task as a point, and the arrows between them show which tasks need to be done before others. We do this to make sure there are no loops or circular dependencies, so it's like a map of our project.

Counting Prerequisites: For each task on our map, we figure out how many prerequisites it has. Prerequisites are other tasks that need to be done before this one. This helps us understand what has to happen before a task can begin.

Putting Things in Order: We organize our tasks in a specific order. We start with tasks that don't have any prerequisites because they can begin right away. We add these to our list in that order. Then, as we finish each task, we update our map to show that its prerequisites are done. We keep doing this until we've put all the tasks in order.

Optimal Sequence: The order we end up with is the best way to complete our project. It's arranged so that tasks with fewer prerequisites or tasks that don't take as long are done first. This helps us finish the entire project as quickly as possible.

The efficiency of this algorithm relies on the topological sorting technique. Here's an analysis of its efficiency:

```python
class TreeNode:
    def __init__(self, identifier):
        self.identifier = identifier
        self.left = None
        self.right = None

def classify_binary_tree(root):
    def traverse(node):
        nonlocal nodes_with_two_children, leaves, total_nodes

        if node is None:
            return

        total_nodes += 1

        if node.left is None and node.right is None:
            # Node has no children, it's a leaf
            leaves += 1
        elif node.left is not None and node.right is not None:
            # Node has two children
            nodes_with_two_children += 1

        # Recursively traverse left and right subtrees
        traverse(node.left)
        traverse(node.right)

    # Initialize counts
    nodes_with_two_children = 0
    leaves = 0
    total_nodes = 0

    # Start traversal from the root
    traverse(root)

    # Determine if the tree is balanced or unbalanced
    if nodes_with_two_children == (leaves - 1):
        classification = "Balanced"
    else:
        classification = "Unbalanced"

    # Return classification and structural information
    return {
        "Classification": classification,
        "Nodes_With_Two_Children": nodes_with_two_children,
        "Leaves": leaves,
        "TotalNodes": total_nodes,
    }

# Example Usage:
# Construct a sample binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)

# Classify the binary tree and get structural information
result = classify_binary_tree(root)
print(result)
```

Given an array of integers, we need to find and count all unique arithmetic subsequences in the array. An arithmetic subsequence is a sequence of elements where the difference between any two consecutive elements is the same.

You can solve this problem efficiently using dynamic programming. Create a dictionary to store the count of arithmetic subsequences ending at each element.
```python
def countArithmeticSubsequences(arr):
    count = 0
    dp = [{} for _ in range(len(arr))]

    for i in range(len(arr)):
        for j in range(i):
            diff = arr[i] - arr[j]
            if diff in dp[j]:
                if diff in dp[i]:
                    dp[i][diff] += dp[j][diff]
                else:
                    dp[i][diff] = dp[j][diff]
                count += dp[j][diff]  # Update the count

            # If the diff is not in dp[i], initialize it to 1
            if diff not in dp[i]:
                dp[i][diff] = 1

    return count

# Example usage:
arr = [2, 4, 6, 8, 10]
result = countArithmeticSubsequences(arr)
print(result)  # Output: 4
```

# Solution:

In this problem, the task is to find the shortest path between two nodes, A and B, in an undirected graph. This problem can be solved efficiently using the Breadth-First Search (BFS) algorithm. BFS explores the graph level by level, ensuring that the first path it finds from A to B is the shortest.

Breadth-First Search (BFS) is a graph traversal and search algorithm that explores the vertices of a graph in breadthward motion. It starts from a specified source vertex and explores all its neighbors before moving to their neighbors, and so on, until all vertices that are reachable from the source have been visited.

Key characteristics of BFS:

Queue-Based: BFS uses a queue data structure to keep track of the vertices to be visited. It starts by enqueueing the source vertex and then dequeues vertices in a first-in-first-out (FIFO) order.

Level by Level: BFS explores vertices level by level, making it suitable for finding the shortest path between two vertices in unweighted graphs.

Used for Traversal and Shortest Paths: BFS can be used to traverse and search for specific vertices in a graph. It's particularly efficient for finding the shortest path in an unweighted graph.

Optimal for Shortest Paths: When applied to unweighted graphs, BFS guarantees that the first path it finds from the source to any other vertex is the shortest path.

Applications: BFS has various applications in computer science and real-world scenarios, such as web crawling, network routing, shortest path problems, and more

Here's a pseudocode for the BFS algorithm to find the shortest path from node A to node B in an undirected graph:

function bfs_shortest_path(graph, start, end):

The Reconstruct_Path function is used to reconstruct the shortest path from the parent dictionary.

Here are the steps to determine the shortest path from node A to node B using Breadth-First Search (BFS) in a graph:

1) Start with an empty queue for BFS traversal.
2) Create a visited set or array to keep track of visited nodes.
3) Create a parent dictionary or array to store the parent node of each vertex in the shortest path.
4) Enqueue the source node A into the queue and mark it as visited.
   While the queue is not empty:
   a. Dequeue the front node from the queue.
   b. If the dequeued node is the destination node B, you have found the shortest path, and you can exit the loop.
   c. Explore all neighbors of the current node that have not been visited:
      Mark the neighbor as visited.
      Set the parent of the neighbor to be the current node.
      Enqueue the neighbor into the queue for further exploration.
6) If you reach this point and the queue becomes empty, there is no path from A to B.
7) To reconstruct the shortest path from A to B, use the parent dictionary:
   Start from node B as the current node.
   Backtrack from B to A by following the parent pointers.
   Add each node to the path to maintain the correct order.
   
These steps define the process of using BFS to find the shortest path between two nodes in a graph.

```python
# Code:

from collections import deque

class Graph:
    def __init__(self, vertices):
        self.graph = {}
        for vertex in vertices:
            self.graph[vertex] = []
    
    def addEdge(self, u, v, weight):
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))

    def bfs_shortest_path(self, start, end):
        # Initialize the visited set and the queue for BFS
        visited = set()
        queue = deque([(start, [start])])

        while queue:
            current_vertex, path = queue.popleft()
            visited.add(current_vertex)

            for neighbor, weight in self.graph[current_vertex]:
                if neighbor not in visited:
                    new_path = path + [neighbor]

                    # If the destination vertex is reached, return the path
                    if neighbor == end:
                        return new_path

                    queue.append((neighbor, new_path))

        return None  # No path found

# Example usage:
vertices = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B', 2), ('A', 'C', 3), ('B', 'D', 5), ('C', 'E', 4), ('D', 'E', 1)]

graph = Graph(vertices)
for edge in edges:
    graph.addEdge(edge[0], edge[1], edge[2])

source = 'A'
destination = 'E'
shortest_path = graph.bfs_shortest_path(source, destination)
print("Shortest Path:", shortest_path)
```

To solve this problem, I will use Dijkstra's algorithm to find the shortest path from the "Entrance" chamber to the "Treasure" chamber while avoiding traps. Dijkstra's algorithm works by iteratively selecting the vertex with the smallest known distance and relaxing its neighbors.

Dijkstra's algorithm is a greedy algorithm used to find the shortest path in a weighted graph. It guarantees finding the shortest path when all edge weights are non-negative. In this problem, we modify Dijkstra's algorithm to handle traps by avoiding chambers with traps.

The priority queue ensures that we always explore the chamber with the shortest known distance first. By relaxing edges and considering traps, we find the shortest time to reach the "Treasure" chamber while avoiding any dangerous traps.

Here are the steps to find the shortest time:

1. Initialize a priority queue (min-heap) to store chambers based on their minimum travel times from the "Entrance" chamber. Initialize distances to all chambers as infinity except for the "Entrance," which is set to 0.

2. While the priority queue is not empty:
   - Dequeue the chamber with the smallest known distance (let's call it `current_chamber`).
   - If `current_chamber` is the "Treasure" chamber, return its distance as the shortest time.
   
   **Note:** If you encounter a chamber with a trap, skip it and continue the loop.

3. For each neighbor (`neighbor_chamber`) of `current_chamber`:
   - Calculate the tentative distance to `neighbor_chamber` through `current_chamber`. If there's a trap on the edge, multiply the travel time by 2.
   - If the tentative distance is less than the current distance to `neighbor_chamber`, update the distance and enqueue `neighbor_chamber` with its new distance.

4. If you finish the loop without finding a path to the "Treasure" chamber, return -1 to indicate that there is no safe path.

function shortestPathWithTraps(graph, A, B, traps):
    distances = {}  # Dictionary to store minimum distances to nodes
    priorityQueue = MinHeap()  # Min-heap to maintain nodes with the shortest distance
    visited = set()  # Set to track visited nodes

    distances[A] = 0
    
    # Enqueue A with its distance to start the algorithm
    priorityQueue.enqueue(A, 0)
    
    while priorityQueue is not empty:
        current, current_distance = priorityQueue.dequeue()
        
        # Skip if this node is a trap
        if current in traps:
            continue
        
        # Mark current node as visited
        visited.add(current)
        
        # If we reached node B, return its distance as the shortest path
        if current == B:
            return current_distance
        
        # Explore neighbors of the current node
        for neighbor, weight in graph[current]:
            # Skip neighbors that are traps
            if neighbor in traps:
                continue
            
            # Calculate the tentative distance through the current node
            tentative_distance = current_distance + weight
            
            # If the tentative distance is shorter, update the distance
            if tentative_distance < distances[neighbor]:
                distances[neighbor] = tentative_distance
                priorityQueue.enqueue(neighbor, tentative_distance)
    
    # If we couldn't reach B, return -1

```python
import heapq

def shortestPathWithTraps(graph, start, end, traps):
    # Initialize distances and visited nodes
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    visited = set()

    # Create a priority queue to select nodes with the shortest distance
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current = heapq.heappop(priority_queue)

        if current in visited:
            continue

        if current == end:
            return current_distance

        visited.add(current)

        for neighbor, weight in graph[current]:
            if neighbor not in traps:
                distance = current_distance + weight

                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(priority_queue, (distance, neighbor))

    return float('inf')

# Define the graph
graph = {
    'Entrance': [('A', 3), ('C', 1)],
    'A': [('B', 2)],
    'B': [('Treasure', 2)],
    'C': [('B', 1)],
    'Treasure': []
}

# Define the set of trap nodes
traps = {'A'}

# Find the shortest time from 'Entrance' to 'Treasure' while avoiding traps
shortest_time = shortestPathWithTraps(graph, 'Entrance', 'Treasure', traps)
print("Shortest Time To Treasure:", shortest_time,"minutes")
```
