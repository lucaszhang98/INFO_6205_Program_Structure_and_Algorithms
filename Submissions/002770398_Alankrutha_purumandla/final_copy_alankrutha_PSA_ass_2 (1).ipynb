What is a Minimum spanning Tree?

Given a Graph, the MST of a graph is a subgraph such that all the vertices of the grapgh are included , connects all vertices , with no cycles and the sum of weights is as minimum as possible.


Kruskal's Algorithm:

1. Initialize a forest T, where each vertex in the graph is a separate tree.
2. Create a set S containing all the edges in the graph.
3. Sort the edges in set S in ascending order based on their weights.

Main Loop:
4. While S is not empty and T is not yet spanning (i.e., the number of trees in T is greater than 1):
a. Remove an edge with the minimum weight from set S.
b. If that edge connects two different trees in the forest T, add it to the forest by combining the two trees into a single tree.
c. If the edge connects two vertices within the same tree, discard that edge.

Result:
5. Once the loop terminates, T represents the Minimum Spanning Tree (MST) of the graph.

This algorithm finds the MST by iteratively adding edges with the smallest weights while ensuring that no cycles are formed within the MST. The result is a tree that spans all vertices with the minimum possible total edge weight.

**Topological Sort - Pseudocode**

```python
def topologicalSort(graph):
    result = []
    def dfs(node):
        visited[node] = True
        for neighbor in graph[node]:
            if not visited[neighbor]:
                dfs(neighbor)
        result.append(node)
    visited = {}
    for node in graph:
        visited[node] = False
    for node in graph:
        if not visited[node]:
            dfs(node)
    # Reverse the result to obtain the topological ordering
    result.reverse()
    return result
```

Question 5: Suppose we have a knapsack which can hold int w = 10 weight units. We have a total of int n = 4 items to choose from, whose values are represented by an array int[] val = {10, 40, 30, 50} and weights represented by an array int[] wt = {5, 4, 6, 3}.

Since this is the 0–1 knapsack problem, we can either include an item in our knapsack or exclude it, but not include a fraction of it, or include it multiple times.

<b>Solution

```python
def knapsack(weights, values, W, n):
    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]


    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    max_value = dp[n][W]

    selected_items = []
    i, w = n, W
    while i > 0 and w > 0:
        if dp[i][w] != dp[i - 1][w]:
            selected_items.append(i - 1)
            w -= weights[i - 1]
        i -= 1

    return max_value, selected_items


values = [10, 40, 30, 50]
weights = [5, 4, 6, 3]
W = 10
n = len(values)

max_value, selected_items = knapsack(weights, values, W, n)

print("Maximum value:", max_value)
print("Selected items:", selected_items)
```

  <h1>Question 7: Finding a Recipe in a Cookbook</h1>

  <p>Your friend is an avid cook, and they have a vast collection of cookbooks. They recently acquired a new cookbook containing a variety of recipes. However, they are overwhelmed by the sheer number of recipes in the book and want your help in finding a specific recipe efficiently.</p>

  <p>Each recipe in the cookbook is represented by two arrays of strings:</p>
  <ol>
      <li>An array of strings, <code>cookbookIngredients[]</code>, representing the list of ingredients in each recipe.</li>
      <li>An array of strings, <code>cookbookInstructions[]</code>, representing the cooking instructions for each recipe.</li>
  </ol>

  <p>Your friend wants to know if a particular recipe they are looking for is present in the cookbook. They'll provide you with two arrays of strings representing the recipe they want:</p>

  <ol>
      <li>An array of strings, <code>recipeIngredients[]</code>, representing the list of ingredients for the desired recipe.</li>
      <li>An array of strings, <code>recipeInstructions[]</code>, representing the cooking instructions for the desired recipe.</li>
  </ol>

  <p>Your task is to determine if the provided recipe (list of ingredients and instructions) is present in the cookbook. To be considered a match, both the ingredients and instructions must match exactly, but they don't have to be in the same order.</p>

  
  <p>Your function should return <code>true</code> if the recipe with both the same ingredients and instructions is found in the cookbook; otherwise, it should return <code>false</code>.</p>


<b>Solution</b><br>
The C++ function findRecipe takes two arrays of strings representing cookbook ingredients and instructions, as well as the lengths of these arrays (denoted as n for the cookbook and m for the recipe). It verifies the inputs for non-negativity and then iterates through the cookbook arrays, comparing elements with corresponding elements in the recipe arrays. When matching ingredients and instructions are found, it increments counters. Finally, it checks if both counters are equal to m, indicating the recipe's presence in the cookbook with the same ingredients and instructions in the same order. The function returns true in this case; otherwise, it returns false. The time complexity of this solution is O(n), where n is the number of recipes in the cookbook.

```python
bool findRecipe(const std::string cookbookIngredients[], const std::string cookbookInstructions[], int n, const std::string recipeIngredients[], const std::string recipeInstructions[], int m) {
    if (n < 0 || m < 0) {
        return false;
    }

    int ingredientIndex = 0;
    int instructionIndex = 0;

    for (int recipeIndex = 0; recipeIndex < n && (ingredientIndex < m || instructionIndex < m); recipeIndex++) {
        if (cookbookIngredients[recipeIndex] == recipeIngredients[ingredientIndex]) {
            ingredientIndex++;
        }
        if (cookbookInstructions[recipeIndex] == recipeInstructions[instructionIndex]) {
            instructionIndex++;
        }
    }

    return (ingredientIndex == m && instructionIndex == m);
}
```
