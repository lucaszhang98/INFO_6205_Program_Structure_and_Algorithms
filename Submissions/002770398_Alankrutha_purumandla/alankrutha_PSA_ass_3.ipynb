# Question 1
 Give a brief definition for the following terms:
<ol>
<li> Ford-Fulkerson Algorithm
<li> Push-Relabel Algorithm
<li> Residual Graph
<li> Augmenting Path
</ol>

<b>Solution:</b>
<ol>
<br>
<li>
Ford-Fulkerson Algorithm:<br>
The Ford-Fulkerson algorithm is a method for finding the maximum flow in a flow network. It repeatedly augments the flow along augmenting paths until no more such paths can be found.
<br>
<li> Push-Relabel Algorithm:<br>
The Push-Relabel algorithm is another approach for finding the maximum flow in a network. It operates by maintaining a preflow and iteratively performing "push" and "relabel" operations on nodes to improve the flow.
<br>
<li>
Residual Graph:<br>
In the context of network flow algorithms, a residual graph is a modified graph representing the remaining capacity on the edges after some flow has been sent. It's used to keep track of available capacity for further augmenting paths.
<br>
<li>Augmenting Path:<br>
An augmenting path is a path in a flow network from the source to the sink along which additional flow can be sent. It's used in the context of network flow algorithms to increase the total flow in the network.

You are allowed to delete up to k edges from the graph.
<br><br>
Find a polynomial-time algorithm to determine which edges to delete to minimize the number of edge deletions while ensuring that vertices i and j are no longer connected in the remaining graph.

<b>Solution:</b>
<br>
We can solve this problem by utilizing the concept of edge connectivity. Edge connectivity represents the minimum number of edges that must be removed to disconnect two vertices. In our case, we want to disconnect vertices i and j.
<br>
<li>If the edge connectivity between i and j is already less than or equal to k, then we can achieve the desired outcome by deleting the edges that belong to any minimum cut separating i and j. There is no need to delete more edges.
<li>If the edge connectivity between i and j is greater than k, we need to delete additional edges. Find a minimum cut separating i and j in the graph G.
<li>Delete any k edges from the minimum cut to ensure that there are k or fewer edges connecting i and j in the remaining graph.
<br>
<br>
This solution guarantees that vertices i and j are disconnected with a minimum number of edge deletions. The overall time complexity of this algorithm depends on the algorithm used for edge connectivity and minimum cut computations, which are typically polynomial for most practical graphs.

Each resident should be assigned to a shelter within a half-hour's driving distance of their current location.
<br><br>
The city's emergency management team needs to determine if it's possible to assign residents to shelters in a way that balances the load on the shelters, ensuring that no shelter is overloaded.
<br><br>
Provide a polynomial-time algorithm to decide whether it is possible to assign residents to shelters in a balanced manner, taking into account their locations and the shelter capacities.

<b>Solution</b>
<li>Create a flow network with a source node 'S' and a sink node 'T'.
<li>Create a node for each resident, representing their current location. Connect each resident node to the source node 'S' with an edge of capacity 1 (each resident needs to be sent to a shelter).
<li>Create a node for each emergency shelter, and connect each shelter node to the sink node 'T' with an edge of capacity [n/m] (the maximum number of residents each shelter can receive).
<li>Create edges from each resident node to each shelter node with a capacity of 1 if the driving time from the resident's location to the shelter is within half an hour, and 0 otherwise.
<li>Run the Ford-Fulkerson algorithm to find the maximum flow in the network from 'S' to 'T'.
<li>If the maximum flow is equal to 'n' (the total number of residents), then it is possible to assign residents to shelters in a balanced way. Otherwise, it is not possible.

The algorithm's time complexity is determined by the Ford-Fulkerson algorithm, which is generally polynomial in the number of nodes and the capacity of the edges. In this case, the number of nodes is O(n + m + 2), and the capacity of the edges is bounded by O(n), so the overall time complexity is polynomial in the size of the problem.

# Question 9
Suppose we have a knapsack which can hold int w = 10 weight units. We have a total of int n = 4 items to choose from, whose values are represented by an integer array <code>val = [15, 30, 25, 50, 10]</code> and weights represented by an another integer array <code> wt = [5, 10, 5, 20, 3]</code>.

Since this is a 0–1 knapsack problem, we can either include an item in our knapsack or exclude it, but not include a fraction of it, or include it multiple times.

```python
def knapsack(weights, values, W, n):
    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]


    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    max_value = dp[n][W]

    selected_items = []
    i, w = n, W
    while i > 0 and w > 0:
        if dp[i][w] != dp[i - 1][w]:
            selected_items.append(values[i - 1])
            w -= weights[i - 1]
        i -= 1

    return max_value, selected_items


values = [15, 30, 25, 50, 10]
weights = [5, 10, 5, 20, 3]
W = 10
n = len(values)

max_value, selected_items = knapsack(weights, values, W, n)

print("We used items", sorted(selected_items), "for a combined value of", max_value, "in the knapsack.")
```
