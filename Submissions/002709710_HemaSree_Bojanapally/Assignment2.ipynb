Which of the following is used for graphs with greater density for implementing Prim's Algorithm efficiently.

A) d-ary heap
B) linear search
C) fibonacci heap
D) binary search

**Answer:** A

**Explanation:**

In Prim's algorithm, selecting the minimum weight edge for the current vertex involves searching through an array of weights. For dense graphs, this search operation can be made efficient by employing a binary heap. Additionally, when dealing with graphs that have high density, the algorithm's performance can be significantly improved by using a d-ary heap, which is an extension of the binary heap. This optimization allows Prim's algorithm to run in linear time, ensuring swift processing even for densely connected graphs.

**Output:**

A set of edges forming the minimum spanning tree

<br>. 1 <= V<= 10<sup>3</sup>
<br>. 0 <= E <= V*(V-1) /2
<br>. 1 <= weight of edges <= 10<sup>3</sup>

**Solution and Justification:**

Kruskal's algorithm involves sorting edges by weight and adding them to the minimum spanning tree if they don't form cycles. The algorithm utilizes disjoint-set data structures.

**Below is the Pseudocode for the Algorithm:**

Step 1: Initially, establish a forest F where each individual vertex within the graph is considered as a separate tree.

Step 2: Create a set E containing all the edges present in the graph.

Step 3: Repeat Steps 4 and 5 as long as the set E is not empty and the forest F is not yet a spanning tree.

Step 4: Select an edge with the smallest weight from the set E.

Step 5: If the edge chosen in Step 4 connects two distinct trees within the forest F, incorporate it into the forest by combining these two trees into one. If the edge does not connect different trees, discard it.

Step 6: Terminate the process as the minimum spanning tree has been constructed.


**Steps:**

I. Connect A-B (1)

II. Connect D-E (2)

III. Connect B-C (3)

IV. Connect C-D (4)

V. Skip A-E (5) (Forms cycle)

VI. Skip A-C (7) (Forms cycle)

VII. Skip A-D (10) (Forms cycle)

Stop MST is now formed (4 edges, 5 vertices)

Cost of Minimum Spanning Tree(MST) is : AB+DE+BC+CD = 1+2+3+4 = 10


**Time Complexity:**

The time complexity of Kruskal's algorithm is O(E logE) or O(V logV), where E is the no. of edges, and V is the no. of vertices.

**Problem Statement:**

Given a list of positive integers representing item prices in a store, and a budget B, find the maximum number of items you can purchase without exceeding the budget using dynamic programming.


**Input:**

A list of positive integers, prices, where each price is between 1 and 1000.

A positive integer, B, representing the budget.

The length of prices is between 1 and 100.

**Output:**

Return an integer indicating the maximum number of items that can be bought within the given budget.


**Sample Inputs and Outputs:**

Input: prices = [30, 20, 50, 40, 60], B = 100.  
Output: 3

Input: prices = [10, 25, 15, 35], B = 60.  
Output: 4

**Constraints:**

Prices in the list can have duplicates.

The function must handle large inputs efficiently.

**Solution and Explanation: **

def max_items_within_budget(prices, B):
    n = len(prices)
    dp = [0] * (B + 1)
    
    for i in range(n):
        for j in range(B, prices[i] - 1, -1):
            dp[j] = max(dp[j], dp[j - prices[i]] + 1)
    
    return dp[B]

A positive integer num (1 ≤ num ≤ 10^8) consisting only of digits 6 and 9.

**Output:**

An integer representing the maximum number you can obtain by changing at most one digit.

**Sample Input and Output:**

Input: num = 6666      
Output: 9666                                                                   
Explanation: Changing the first digit results in 9666. Changing any other digit results in a number smaller than 9666. The maximum number is 9666.

**Constraints:**

The input number num consists only of digits 6 and 9.

**Solution and Justification:**

    def maximum_number(num):
    num_str = str(num)
    for i in range(len(num_str)):
        if num_str[i] == '6':
            return int(num_str[:i] + '9' + num_str[i+1:])
    return num
