A starting node label (an integer) from which the DFS traversal should begin.

**Output:**

**Constraints:**

The input graph is a simple undirected graph.

The input graph has at most 10^4 nodes.

Nodes are labeled with unique positive integers.

The starting node label is guaranteed to be a valid node label in the graph.

**Solution and Justification:**

    def dfs(graph, start_node): visited = set() result = []

    def dfs_recursive(node):
        visited.add(node)
        result.append(node)

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                dfs_recursive(neighbor)

    dfs_recursive(start_node)

    return result

The dfs function takes the graph and a starting node label as input and performs a Depth First Search traversal starting from the given node. It uses a recursive helper function dfs_recursive to explore the graph and adds visited nodes to the result list. Finally, it returns the list of reachable nodes.

**Problem Statement:*

You have a set of tasks, each having a start time and an end time. Your objective is to determine the maximum number of tasks you can complete without any time overlaps. Employ a greedy algorithm to tackle this problem.

**Input:**

A list of tasks, where each task is represented as a tuple (start_time, end_time).
The list is unordered, and tasks may appear in any sequence.

**Output:**

A list of chosen tasks, signifying the highest number of non-overlapping tasks you can accomplish.
The list should be organized in ascending order based on end times.

**Constraints:**

1 <= Number of tasks <= 10^4
0 <= start_time < end_time <= 10^6

**Answer and Explanation:**

Organize the task list by end times in ascending order.
Create an empty list named selected_tasks to store the chosen tasks.
Initialize a variable called last_end_time to 0.
Iterate through the sorted tasks:
If the task's start time is greater than or equal to last_end_time, include it in selected_tasks and update last_end_time with the task's end time.
Return selected_tasks as the final outcome.

**Pseudocode:**

def max_non_overlapping_tasks(tasks):
    tasks.sort(key=lambda x: x[1])
    
    selected_tasks = []
    last_end_time = 0
    
    for task in tasks:
        start_time, end_time = task
        if start_time >= last_end_time:
            selected_tasks.append(task)
            last_end_time = end_time
    
    return selected_tasks
**Sample Input and Output:**

**Input:**


tasks = [(1, 3), (2, 4), (3, 5), (5, 7), (6, 8)]

**Output:**

Result for the provided input would be: [(1, 3), (3, 5), (6, 8)]

**Explanation:**

In the given instance, the greedy algorithm selects tasks in a way that avoids any time overlaps. It starts with the task (1, 3), then chooses (3, 5), and lastly (6, 8). These tasks are non-overlapping and form the largest set of tasks that can be accomplished without overlaps.
