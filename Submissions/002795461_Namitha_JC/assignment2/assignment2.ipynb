**Solution 1:**

Kruskal's Algorithm and Prim's Algorithm are two different algorithms used to find the minimum spanning tree (MST) of a weighted, connected graph. Here are the key differences between the two algorithms:

Kruskal's Algorithm:

Type of Algorithm: Kruskal's algorithm is a greedy algorithm.
Initialization: It starts with an empty set of edges and adds edges to the MST in increasing order of their weights.
Edge Selection: At each step, it selects the minimum-weight edge that does not form a cycle in the MST.
Data Structure: It often uses a disjoint-set data structure (Union-Find) to efficiently check for cycles and maintain the connected components.
Edge Consideration: Kruskal's algorithm considers edges independently and does not rely on a starting vertex.
Time Complexity: The time complexity of Kruskal's algorithm is typically O(E log E), where E is the number of edges in the graph. It depends on sorting the edges by weight.

Prim's Algorithm:

Type of Algorithm: Prim's algorithm is also a greedy algorithm.
Initialization: It starts with an arbitrary vertex and grows the MST by adding the minimum-weight edge that connects a vertex in the tree to a vertex outside the tree.
Edge Selection: At each step, it selects the minimum-weight edge that connects a vertex in the MST to a vertex outside the MST.
Data Structure: Prim's algorithm often uses a priority queue (or heap) to efficiently select the minimum-weight edges.
Edge Consideration: Prim's algorithm depends on a starting vertex and grows the MST from there.
Time Complexity: The time complexity of Prim's algorithm is typically O(E + V log V), where E is the number of edges and V is the number of vertices in the graph. It depends on selecting edges efficiently using a priority queue.

**Solution 4:**


Assumption made: the intervals (or tasks) are already sorted by end time:

Given intervals (or tasks) and their associated values:

```
Task     End Time  Value
------   --------  -----
Task 1   2         6
Task 2   4         3
Task 3   5         4
Task 4   3         5
Task 5   6         3
```

For each task, we'll find the "previous non-overlapping task" (if any). Then, we'll compute the maximum combined value for the set of tasks up to and including that task.

Here's the dynamic programming table:

```
Task     Value  Previous  Max
------   -----  --------  -----------------------------
Task 1   6      n/a       Max(6+0, 0) = 6
Task 4   5      n/a       Max(5+0, 6) = 6
Task 2   3      Task 1    Max(3+6, 6) = 9
Task 3   4      Task 4    Max(4+6, 9) = 10
Task 5   3      Task 2    Max(3+9, 10) = 12
```

Now, let's backtrack to determine the optimal set of tasks:

```
Task     Trace(i)                 S
------   ----------------------   ---------------
Task 5   3 + 9 = 12, jump to Task 2 {Task 5}
Task 3   4 + 6 = 10, jump to Task 4 {Task 5}
Task 2   3 + 6 = 9, jump to Task 1 {Task 5, Task 2}
Task 4   5 = 5, stay at Task 4    {Task 5, Task 2}
Task 1   6 = 6, stay at Task 1    {Task 5, Task 2}
```

Thus, the optimal set of tasks is S = {Task 2, Task 5} with a combined value of 12.

Pseudo Code:



```
Function findOptimalTaskSet(tasks):
    Sort tasks by end time in ascending order

    n = length(tasks)
    dp = [0] * n

    # Initialize the dynamic programming table
    dp[0] = tasks[0].value

    For i from 1 to n-1:
        value = tasks[i].value
        endTime = tasks[i].endTime

        # Find the latest non-overlapping task
        j = i - 1
        While j >= 0 and tasks[j].endTime >= endTime:
            j = j - 1

        # Update the dynamic programming table
        If j >= 0:
            value = value + dp[j]

        dp[i] = maximum(value, dp[i-1])

    # Backtrack to find the optimal task set
    optimalTaskSet = []
    i = n - 1
    While i >= 0:
        If i == 0 and dp[i] > 0:
            Add tasks[i] to optimalTaskSet
            Break
        ElseIf i > 0 and dp[i] > dp[i-1]:
            Add tasks[i] to optimalTaskSet
            i = i - 2
        Else:
            i = i - 1

    Reverse optimalTaskSet

    Return optimalTaskSet, dp[n-1]
```

**Solution 5:**


---

**Algorithm**: Dynamic Programming 0/1 Knapsack

**Input**:
- `values[]`: Array containing values of the items
- `weights[]`: Array containing weights of the items
- `W`: Maximum weight the knapsack can hold
- `n`: Number of items

**Output**:
- Maximum total value the knapsack can hold without exceeding weight `W`
- List of selected items

---

**Steps**:

1. **Initialization**:
   - Create a 2D array `dp` of dimensions `(n+1) x (W+1)`, initialized to zero.

2. **Base Case**:
   - For `i = 0` to `n`: `dp[i][0] = 0` // For zero capacity, the maximum value is zero
   - For `w = 0` to `W`: `dp[0][w] = 0` // For zero items, the maximum value is zero

3. **Populate the DP table**:
   - For `i = 1` to `n`:
     - For `w = 1` to `W`:
       - If `weights[i-1] <= w`: // If the item can be included in the current weight
           - `dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])` // Choose the max value between including and excluding the item
       - Else:
           - `dp[i][w] = dp[i-1][w]` // Item can't be included, so value remains the same as without this item

4. **Backtrack to find selected items**:
   - Initialize an empty list `items_selected`
   - Initialize `w = W`
   - For `i = n` down to `1`:
     - If `dp[i][w]` is not equal to `dp[i-1][w]`: // This item was included
       - Append `i-1` (item index) to `items_selected`
       - Decrement `w` by `weights[i-1]` // Adjust weight for next iteration

5. **Return Result**:
   - Return `dp[n][W]` (Maximum value) and `items_selected`

---

This algorithm provides a bottom-up solution to the 0/1 Knapsack problem using dynamic programming, offering optimal performance for the problem.

Pseudo Code:



```
Function Knapsack(values, weights, W):
    n = length(values)
    Create a table dp of size (n+1) x (W+1) initialized to zero

    // Initialization
    FOR w FROM 0 TO W:
        dp[0][w] = 0
    FOR i FROM 0 TO n:
        dp[i][0] = 0

    // Fill the table based on values and weights
    FOR i FROM 1 TO n:
        FOR w FROM 1 TO W:
            IF weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
            ELSE:
                dp[i][w] = dp[i-1][w]

    // Trace back to find the selected items
    items_selected = empty list
    w = W
    FOR i FROM n DOWN TO 1:
        IF dp[i][w] != dp[i-1][w]:
            append i-1 to items_selected
            w = w - weights[i-1]

    RETURN dp[n][W], items_selected

```

In a binary tree, let's define a "full node" as a node with two children and a "half node" as a node with exactly one child. Show by induction that in any non-empty binary tree, the number of full nodes plus one is at least the number of half nodes.

**Solution 7:**

**Base Case**: Consider a binary tree with just 1 node (root). In this scenario, there are 0 full nodes and 0 half nodes. Our claim is \(0 + 1 \geq 0\), which holds true.

**Inductive Step**: Assume the claim is true for a binary tree with \(n\) nodes, where \(F\) is the number of full nodes and \(H\) is the number of half nodes, so \(F + 1 \geq H\).

Now, let's add a new node to this tree:

1. If we add this node as a leaf to a current leaf, it will make the current leaf a half node. This results in \(H' = H + 1\), but \(F' = F\) (since no new full nodes were created). Thus, the claim still holds: \(F' + 1 = F + 1 \geq H + 1 = H'\).

2. If we add this node as the second child to a half node, it will convert the half node into a full node. This results in \(H' = H - 1\) and \(F' = F + 1\). Here, \(F' + 1 = F + 2 > H\), so the claim still holds: \(F' + 1 > H'\).

Given our base case and inductive step, it follows by induction that for any non-empty binary tree, the number of full nodes plus one is at least the number of half nodes.

**Pseudo code** for counting Full and Half nodes:

```
function countNodes(node):
    if node is NULL:
        return (0, 0) # Returns (full nodes, half nodes)

    leftFull, leftHalf = countNodes(node.leftChild)
    rightFull, rightHalf = countNodes(node.rightChild)

    totalFull = leftFull + rightFull
    totalHalf = leftHalf + rightHalf

    # Check the current node
    if node has two children:
        totalFull += 1
    elif node has one child:
        totalHalf += 1

    return (totalFull, totalHalf)

# Use the function on the root of the tree
```

**Solution 8:**

Approach:

Initialize two pointers: one for S' (let's call it ptr_s) and one for S (let's call it ptr_s_long).

Iterate through both sequences S' and S simultaneously, advancing both pointers based on whether the current word in S matches the current word in S'.

If the current word in S matches the current word in S', move both pointers forward.

If the current word in S doesn't match the current word in S', only move the pointer in S forward.

Continue this process until you reach the end of S' or the end of S.

If you reach the end of S' while iterating, it means that S' is a subsequence of S. Return true.

If you reach the end of S while iterating, it means that S' is not a subsequence of S. Return false.

Pseudo code:



```
function isSubsequence(S', S):
    ptr_s = 0  # Pointer for S'
    ptr_s_long = 0  # Pointer for S

    while ptr_s < len(S') and ptr_s_long < len(S):
        if S'[ptr_s] == S[ptr_s_long]:
            ptr_s += 1
        ptr_s_long += 1

    # Check if we reached the end of S'. If yes, S' is a subsequence of S.
    return ptr_s == len(S')
```

Code:

````python
def isSubsequence(S, S_prime):
    ptr_s = 0  # Pointer for S
    ptr_s_prime = 0  # Pointer for S'

    while ptr_s < len(S) and ptr_s_prime < len(S_prime):
        if S_prime[ptr_s_prime] == S[ptr_s]:
            ptr_s_prime += 1
        ptr_s += 1

    # Check if we reached the end of S'. If yes, S' is a subsequence of S.
    return ptr_s_prime == len(S_prime)

# Example usage:
S = "This document contains important information that you need to read. The information is very important."
S_prime = "important information"

result = isSubsequence(S.split(), S_prime.split())  # Splitting the strings into lists of words

if result:
    print("S' is a subsequence of S.")
else:
    print("S' is not a subsequence of S.")
````

**Question 9:**

Given below is a Weighted Directed Graph, apply BFS, DFS and compare the differnces between them.

dfs.png

**Solution 9:**

Code Implementation for BFS and DFS

````python
graph = {
    'A': {'B': 2, 'C': 3},
    'B': {'D': 1, 'E': 4},
    'C': {'F': 5},
    'D': {},
    'E': {'F': 6},
    'F': {}
}
````

````python
#BFS

from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            result.append(vertex)
            queue.extend(node for node in graph[vertex] if node not in visited)
    return result

print(bfs(graph, 'A'))
````

````python
#DFS

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited

print(dfs(graph, 'A'))
````

Differences between BFS and DFS in the context of directed graphs (the basics remain the same):

Traversal Order:

BFS visits all the immediate descendants before moving to the next level.
DFS explores as deep as possible along one lineage before backtracking.
Pathfinding:

BFS can be used for finding the shortest path in unweighted directed graphs.
DFS is not commonly used for pathfinding in directed graphs as it can traverse longer paths unnecessarily.
Cycle Detection:

In directed graphs, DFS can be modified to detect cycles, making it useful for checking properties like acyclicity.
BFS is not commonly used for cycle detection in directed graphs.
Implementation:

BFS uses a queue (FIFO) data structure.
DFS can be implemented using recursion or an explicit stack (LIFO) data structure.
Memory:

BFS can be more memory-intensive in worst cases since it might need to store all neighbors at the current level.
DFS usually uses less memory as it needs to store only the nodes along the current path.

The time complexity for both BFS and DFS in terms of graph traversal remains O(V + E), where V is the number of vertices (nodes) and E is the number of edges, regardless of whether the graph is weighted or unweighted. The addition of weights doesn't change the traversal complexity. However, for directed graphs, the number of edges originating from a vertex can be different from undirected graphs, but the fundamental complexity analysis remains the same.

Here's the reasoning:

Breadth-First Search (BFS):
Vertices: Each vertex is processed once, so the total time for all vertices is O(V).

Edges: In the worst case, every edge is examined once. For a vertex v, all its outgoing edges are examined. In a directed graph, every edge will be examined exactly once. So, the total time for all edges is O(E).

Combining these, the time complexity for BFS is O(V + E).

Depth-First Search (DFS):
Vertices: Similarly, each vertex is processed once, resulting in O(V) time.

Edges: DFS will check all outgoing edges of a vertex to decide where to go next. In a directed graph, each edge will be examined once to decide if the DFS should go down that path or not. Thus, the time for all edges is O(E).

Combining these, the time complexity for DFS is O(V + E).

It's important to note that the O(V) term can often be overshadowed by the O(E) term, especially in dense graphs where the number of edges is much larger than the number of vertices. However, in cases like trees (which are a special type of graph), where E = V - 1, both terms are significant.






**Question 10:**

Find shortest path from node A to node B using Dijkstra's algorithm in a graph of your choice with a graph, code implementation and algorithm.

**Solution 10:**


Steps:

Assign to every node a tentative distance value: set it to zero for our initial node and to infinity for all other nodes.

Set the initial node as current.

Mark all other nodes unvisited. Create a set of all the unvisited nodes called the unvisited set.

For the current node, consider all of its neighbors and calculate their tentative distances.

If the newly calculated tentative distance is less than the current assigned value, update the distance.

After considering all of the neighbors of the current node, mark the current node as visited. A visited node will not be checked again.

Select the unvisited node with the smallest tentative distance, and set it as the new current node. Return to step 3.

Pseudo Code:



```
function Dijkstra(graph, start, end):
    create an empty set called visited
    create a dictionary called distances with all nodes' distances set to infinity
    create a dictionary called predecessors with all nodes' predecessors set to null
    set the distance for the start node to 0
    create a priority queue called queue and add (start, 0) to it

    while queue is not empty:
        current_node, current_distance = get node with the smallest distance from queue

        if current_node is equal to end:
            break

        if current_node is not in visited:
            add current_node to visited

            for neighbor in neighbors of current_node:
                if neighbor is not in visited:
                    calculate tentative_distance = distances[current_node] + distance between current_node and neighbor

                    if tentative_distance < distances[neighbor]:
                        distances[neighbor] = tentative_distance
                        predecessors[neighbor] = current_node
                        add (neighbor, tentative_distance) to queue

    if distances[end] is infinity:
        return "No path found"
    else:
        create an empty list called path
        current_node = end

        while current_node is not null:
            add current_node to the beginning of path
            current_node = predecessors[current_node]

        return path

```



Algorithm:

```
Starting from A:

Distance to A is 0.
Distance to C is 2.
Distance to D is 4.

Distance to D from C is 3 (2 from A to C + 1 from C to D).
Distance to B from C directly is 7 (2 from A to C + 5 from C to B).
The next shortest known distance is to D, so D is our next current node:

Distance to E is 6 (4 from A to D + 2 from D to E).
E has a direct link to B with a distance of 2, so total distance is 8 (6 to E + 2 to B).

B is our target, and we have already reached it. The shortest path is: A -> C -> D -> E -> B.
```
